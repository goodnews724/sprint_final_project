````python
import pandas as pd

# 먼저 필요한 컬럼들로 샘플 데이터 로드
sample = pd.read_csv(gcs_path, nrows=50000)

# session ID에서 해시 부분 추출
sample['session_hash'] = sample['event_properties.sessionid'].str.split('.').str[1]

# 4. 장치별 해시 패턴
device_hash_mapping = sample.groupby('device_id')['session_hash'].nunique()
print(f"장치별 평균 해시 개수: {device_hash_mapping.mean():.2f}")
print(f"장치별 최소 해시 개수: {device_hash_mapping.min()}")
print(f"장치별 최대 해시 개수: {device_hash_mapping.max()}")

# 분포 확인
print("\n장치별 해시 개수 분포:")
print(device_hash_mapping.value_counts().head(10))

# 5. 상세 매핑 테이블 만들기
mapping_analysis = sample.groupby(['session_hash']).agg({
    'user_id': ['nunique', lambda x: list(set(x))[:3]],  # 처음 3명만
    'device_id': ['nunique', lambda x: list(set(x))[:3]], # 처음 3개만
    'event_key': 'count'
}).round(2)

print("\n=== 해시별 상세 매핑 (상위 10개) ===")
print(mapping_analysis.head(10))

# 추가 분석: 해시의 성격 파악
print("\n=== 해시 성격 분석 ===")
hash_stats = mapping_analysis.copy()
hash_stats.columns = ['unique_users', 'user_list', 'unique_devices', 'device_list', 'event_count']

print(f"전체 고유 해시 개수: {len(hash_stats)}")
print(f"단일 사용자 해시: {(hash_stats['unique_users'] == 1).sum()}")
print(f"단일 장치 해시: {(hash_stats['unique_devices'] == 1).sum()}")
print(f"멀티 사용자 해시: {(hash_stats['unique_users'] > 1).sum()}")
print(f"멀티 장치 해시: {(hash_stats['unique_devices'] > 1).sum()}")
````

````python
import pandas as pd
import json
import re

def process_large_json_array():
    json_path = 'work/data/2023-06-17.json'
    csv_path = 'work/data/2023-06-17-converted.csv'
    
    print("청크 단위 JSON 배열 처리 시작...")
    
    chunk_data = []
    chunks_written = 0
    brace_count = 0
    current_object = ""
    in_string = False
    escape_next = False
    
    with open(json_path, 'r', encoding='utf-8') as f:
        # '[' 건너뛰기
        f.read(1)
        
        while True:
            char = f.read(1)
            if not char:
                break
            
            # 문자열 내부 체크
            if char == '"' and not escape_next:
                in_string = not in_string
            elif char == '\\' and in_string:
                escape_next = not escape_next
            else:
                escape_next = False
            
            if not in_string:
                if char == '{':
                    brace_count += 1
                elif char == '}':
                    brace_count -= 1
            
            current_object += char
            
            # 완전한 객체를 찾았을 때
            if brace_count == 0 and current_object.strip():
                # 쉼표 제거
                obj_str = current_object.strip().rstrip(',').strip()
                
                if obj_str and obj_str != ']':
                    try:
                        obj = json.loads(obj_str)
                        chunk_data.append(obj)
                        
                        # 1만개마다 저장 (메모리 절약)
                        if len(chunk_data) >= 10000:
                            df = pd.json_normalize(chunk_data)
                            
                            mode = 'w' if chunks_written == 0 else 'a'
                            header = chunks_written == 0
                            df.to_csv(csv_path, mode=mode, header=header, index=False)
                            
                            chunks_written += 1
                            print(f"청크 {chunks_written} 저장완료 ({len(chunk_data):,}개 레코드)")
                            chunk_data = []
                            
                    except json.JSONDecodeError:
                        pass
                
                current_object = ""
            
            # 진행상황 출력 (매 100MB마다)
            if f.tell() % 100000000 == 0:
                print(f"처리중... {f.tell()//1000000}MB")
    
    # 마지막 청크 저장
    if chunk_data:
        df = pd.json_normalize(chunk_data)
        mode = 'w' if chunks_written == 0 else 'a'
        header = chunks_written == 0
        df.to_csv(csv_path, mode=mode, header=header, index=False)
        chunks_written += 1
        print(f"마지막 청크 저장완료 ({len(chunk_data):,}개 레코드)")
    
    print(f"\n변환 완료! 총 {chunks_written}개 청크 저장됨")
    return csv_path

# 실행
csv_file = process_large_json_array()
````

````python
# 모든 여러 세션 사용자들의 패턴 분석
print("=== 전체 여러 세션 사용자 (2,849명) 패턴 분석 ===")

# 시간 컬럼을 datetime으로 변환
session_data['timestamp'] = pd.to_datetime(session_data['Asia/Seoul'])

all_multi_session_users = multi_session_users['user_id'].tolist()

# 모든 사용자의 세션 패턴 수집
session_patterns = []
session_gaps_all = []
session_durations = []

for user_id in all_multi_session_users:
    user_data = session_data[session_data['user_id'] == user_id]
    sessions = user_data['session_id'].unique()
    
    session_times = []
    session_info = []
    
    for session_id in sessions:
        session_events = user_data[user_data['session_id'] == session_id]
        start_time = session_events['timestamp'].min()
        end_time = session_events['timestamp'].max()
        duration = (end_time - start_time).total_seconds() / 60  # 분 단위
        event_count = len(session_events)
        
        session_times.append(start_time)
        session_info.append({
            'user_id': user_id,
            'session_id': session_id,
            'start_time': start_time,
            'end_time': end_time,
            'duration_minutes': duration,
            'event_count': event_count
        })
        
        session_durations.append(duration)
    
    session_times.sort()
    
    # 세션 간 간격 계산
    for i in range(1, len(session_times)):
        gap_hours = (session_times[i] - session_times[i-1]).total_seconds() / 3600
        session_gaps_all.append(gap_hours)
    
    # 사용자별 패턴 저장
    total_span = (max(session_times) - min(session_times)).total_seconds() / 3600  # 시간
    session_patterns.append({
        'user_id': user_id,
        'session_count': len(sessions),
        'total_span_hours': total_span,
        'avg_gap_hours': total_span / (len(sessions) - 1) if len(sessions) > 1 else 0
    })

# 결과 분석
patterns_df = pd.DataFrame(session_patterns)
print(f"분석 완료: {len(patterns_df)}명의 여러 세션 사용자")

print(f"\n=== 세션 개수별 상세 분포 ===")
for session_count in sorted(patterns_df['session_count'].unique()):
    count = len(patterns_df[patterns_df['session_count'] == session_count])
    print(f"{session_count}개 세션: {count}명")

print(f"\n=== 세션 지속 시간 분석 ===")
import numpy as np
print(f"평균 세션 지속 시간: {np.mean(session_durations):.1f}분")
print(f"중간값 세션 지속 시간: {np.median(session_durations):.1f}분")
print(f"최대 세션 지속 시간: {np.max(session_durations):.1f}분")

print(f"\n=== 세션 간격 분석 ===")
print(f"세션 간격 개수: {len(session_gaps_all)}개")
print(f"평균 세션 간격: {np.mean(session_gaps_all):.1f}시간")
print(f"중간값 세션 간격: {np.median(session_gaps_all):.1f}시간")
print(f"최소 세션 간격: {np.min(session_gaps_all):.2f}시간")
print(f"최대 세션 간격: {np.max(session_gaps_all):.1f}시간")

# 세션 간격 분포
gap_ranges = [
    (0, 1, "1시간 이내"),
    (1, 6, "1-6시간"),
    (6, 24, "6-24시간"),
    (24, 72, "1-3일"),
    (72, 168, "3-7일"),
    (168, float('inf'), "1주일 이상")
]

print(f"\n=== 세션 간격 분포 ===")
for min_h, max_h, label in gap_ranges:
    if max_h == float('inf'):
        count = sum(1 for gap in session_gaps_all if gap >= min_h)
    else:
        count = sum(1 for gap in session_gaps_all if min_h <= gap < max_h)
    percentage = count / len(session_gaps_all) * 100
    print(f"{label}: {count}개 ({percentage:.1f}%)")

print(f"\n=== 사용자별 활동 기간 분석 ===")
print(f"평균 활동 기간: {patterns_df['total_span_hours'].mean():.1f}시간")
print(f"중간값 활동 기간: {patterns_df['total_span_hours'].median():.1f}시간")
print(f"최대 활동 기간: {patterns_df['total_span_hours'].max():.1f}시간")
````

````python
# 세션이 나뉘어지는 구체적인 패턴 분석
print("=== 세션 분할 패턴 상세 분석 ===")

# 1. 짧은 간격으로 나뉜 세션들 (1시간 이내)
short_gap_cases = []
medium_gap_cases = []
long_gap_cases = []

for user_id in all_multi_session_users[:50]:  # 샘플 50명으로 시작
    user_data = session_data[session_data['user_id'] == user_id].sort_values('timestamp')
    sessions = user_data['session_id'].unique()
    
    if len(sessions) == 2:  # 2개 세션인 경우만
        session1_data = user_data[user_data['session_id'] == sessions[0]]
        session2_data = user_data[user_data['session_id'] == sessions[1]]
        
        session1_start = session1_data['timestamp'].min()
        session1_end = session1_data['timestamp'].max()
        session2_start = session2_data['timestamp'].min()
        session2_end = session2_data['timestamp'].max()
        
        gap_hours = (session2_start - session1_end).total_seconds() / 3600
        
        # 세션 종료/시작 이벤트 확인
        session1_last_events = session1_data.tail(3)['event_key'].tolist()
        session2_first_events = session2_data.head(3)['event_key'].tolist()
        
        case_info = {
            'user_id': user_id,
            'gap_hours': gap_hours,
            'session1_end_events': session1_last_events,
            'session2_start_events': session2_first_events,
            'session1_event_count': len(session1_data),
            'session2_event_count': len(session2_data),
            'session1_duration_hours': (session1_end - session1_start).total_seconds() / 3600,
            'session2_duration_hours': (session2_end - session2_start).total_seconds() / 3600
        }
        
        if gap_hours <= 1:
            short_gap_cases.append(case_info)
        elif gap_hours <= 24:
            medium_gap_cases.append(case_info)
        else:
            long_gap_cases.append(case_info)

print(f"분석된 케이스:")
print(f"  1시간 이내: {len(short_gap_cases)}개")
print(f"  1-24시간: {len(medium_gap_cases)}개") 
print(f"  24시간 이상: {len(long_gap_cases)}개")

# 1시간 이내 세션 분할 케이스들 분석
print(f"\n=== 1시간 이내 세션 분할 케이스들 ===")
for i, case in enumerate(short_gap_cases[:10]):
    print(f"\n케이스 {i+1}:")
    print(f"  사용자: {case['user_id']}")
    print(f"  간격: {case['gap_hours']:.2f}시간")
    print(f"  세션1 종료 이벤트: {case['session1_end_events']}")
    print(f"  세션2 시작 이벤트: {case['session2_start_events']}")
    print(f"  세션1 지속시간: {case['session1_duration_hours']:.1f}시간 ({case['session1_event_count']}개 이벤트)")
    print(f"  세션2 지속시간: {case['session2_duration_hours']:.1f}시간 ({case['session2_event_count']}개 이벤트)")

# $session_end -> $session_start 패턴 확인
session_end_start_pattern = 0
app_restart_pattern = 0
no_session_end_pattern = 0

for case in short_gap_cases + medium_gap_cases:
    if '$session_end' in case['session1_end_events'] and '$session_start' in case['session2_start_events']:
        session_end_start_pattern += 1
    elif 'launch_app' in case['session2_start_events']:
        app_restart_pattern += 1
    elif '$session_end' not in case['session1_end_events']:
        no_session_end_pattern += 1

print(f"\n=== 세션 분할 패턴 분류 ===")
print(f"정상 종료→시작 패턴: {session_end_start_pattern}개")
print(f"앱 재시작 패턴: {app_restart_pattern}개") 
print(f"비정상 종료 패턴: {no_session_end_pattern}개")
````

````python
# 더 정확한 세션 분할 원인 분석
print("=== 세션 분할 원인 상세 분석 ===")

# 모든 여러 세션 사용자의 패턴을 정확히 분석
all_session_splits = []

for user_id in all_multi_session_users:
    user_data = session_data[session_data['user_id'] == user_id].sort_values('timestamp')
    sessions = user_data['session_id'].unique()
    
    for i in range(len(sessions) - 1):
        session1_data = user_data[user_data['session_id'] == sessions[i]]
        session2_data = user_data[user_data['session_id'] == sessions[i+1]]
        
        session1_start = session1_data['timestamp'].min()
        session1_end = session1_data['timestamp'].max()
        session2_start = session2_data['timestamp'].min()
        session2_end = session2_data['timestamp'].max()
        
        gap_seconds = (session2_start - session1_end).total_seconds()
        
        # 마지막과 첫 이벤트들
        session1_events = session1_data['event_key'].tolist()
        session2_events = session2_data['event_key'].tolist()
        
        # 분할 유형 분류
        split_type = "unknown"
        
        if gap_seconds < 0:
            split_type = "overlapping_sessions"
        elif gap_seconds <= 60:  # 1분 이내
            if '$session_end' in session1_events and '$session_start' in session2_events:
                split_type = "immediate_restart"
            elif '$session_end' not in session1_events:
                split_type = "crash_restart"
            else:
                split_type = "quick_restart"
        elif gap_seconds <= 3600:  # 1시간 이내
            split_type = "short_break"
        else:
            split_type = "long_break"
        
        all_session_splits.append({
            'user_id': user_id,
            'gap_seconds': gap_seconds,
            'gap_hours': gap_seconds / 3600,
            'split_type': split_type,
            'session1_duration_hours': (session1_end - session1_start).total_seconds() / 3600,
            'session2_duration_hours': (session2_end - session2_start).total_seconds() / 3600,
            'session1_event_count': len(session1_data),
            'session2_event_count': len(session2_data),
            'session1_has_end': '$session_end' in session1_events,
            'session2_has_start': '$session_start' in session2_events,
            'session1_last_event': session1_events[-1] if session1_events else None,
            'session2_first_event': session2_events[0] if session2_events else None
        })

splits_df = pd.DataFrame(all_session_splits)

print(f"총 세션 분할 수: {len(splits_df)}")

# 분할 유형별 분포
print(f"\n=== 세션 분할 유형별 분포 ===")
split_type_counts = splits_df['split_type'].value_counts()
for split_type, count in split_type_counts.items():
    percentage = count / len(splits_df) * 100
    print(f"{split_type}: {count}개 ({percentage:.1f}%)")

# 음수 간격 케이스들 상세 분석
overlapping_cases = splits_df[splits_df['split_type'] == 'overlapping_sessions']
print(f"\n=== 겹치는 세션 케이스들 ({len(overlapping_cases)}개) ===")
for idx, case in overlapping_cases.head(5).iterrows():
    print(f"사용자 {case['user_id'][:8]}... : 간격 {case['gap_hours']:.2f}시간")

# 즉시 재시작 케이스들 분석
immediate_cases = splits_df[splits_df['split_type'] == 'immediate_restart']
print(f"\n=== 즉시 재시작 케이스들 ({len(immediate_cases)}개) ===")
print(f"평균 간격: {immediate_cases['gap_seconds'].mean():.1f}초")
print(f"중간값 간격: {immediate_cases['gap_seconds'].median():.1f}초")

# 크래시 재시작 케이스들
crash_cases = splits_df[splits_df['split_type'] == 'crash_restart']
print(f"\n=== 크래시 추정 케이스들 ({len(crash_cases)}개) ===")
print(f"평균 간격: {crash_cases['gap_seconds'].mean():.1f}초")

# 세션 지속시간 vs 분할 유형
print(f"\n=== 분할 유형별 평균 세션 지속시간 ===")
for split_type in split_type_counts.index:
    type_data = splits_df[splits_df['split_type'] == split_type]
    avg_duration1 = type_data['session1_duration_hours'].mean()
    avg_duration2 = type_data['session2_duration_hours'].mean()
    print(f"{split_type}:")
    print(f"  첫 세션 평균: {avg_duration1:.1f}시간")
    print(f"  둘째 세션 평균: {avg_duration2:.1f}시간")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc # 가비지 컬렉션 모듈 임포트

print("🔍 `hackle_properties`: 하나의 `user_id`가 가질 수 있는 `session_id` 개수 탐구")
print("=" * 80)

# 파일 경로 정의
path_hackle_properties = "gs://sprintda05_final_project/hackle/hackle_properties.parquet"

# 필요한 컬럼만 로드: 'user_id'와 'session_id'
essential_cols_hp = ['user_id', 'session_id']

try:
    print(f"⏳ `hackle_properties` 데이터 로드 중... (컬럼: {essential_cols_hp})")
    df_hp = pd.read_parquet(
        path_hackle_properties,
        columns=essential_cols_hp
    )
    print(f"✅ `hackle_properties` 데이터 로드 완료: 총 {len(df_hp):,}개 행.")

    # 1. 'user_id' 및 'session_id' 컬럼의 결측치 확인 및 처리 (중요)
    # user_id가 비어있는 경우, 해당 세션이 익명 사용자일 가능성이 높습니다.
    # user_id 기준 분석이므로, user_id가 없는 행은 제외하는 것이 합리적입니다.
    user_id_null_count = df_hp['user_id'].isnull().sum()
    session_id_null_count = df_hp['session_id'].isnull().sum()

    print(f"\n📊 `user_id` 결측치 개수: {user_id_null_count:,}개 ({user_id_null_count / len(df_hp) * 100:.2f}%)")
    print(f"📊 `session_id` 결측치 개수: {session_id_null_count:,}개 ({session_id_null_count / len(df_hp) * 100:.2f}%)")

    # user_id가 없는 행은 분석에서 제외 (익명 세션)
    if user_id_null_count > 0:
        df_hp_filtered = df_hp.dropna(subset=['user_id']).copy()
        print(f"✅ `user_id` 결측치 제거 후 남은 행: {len(df_hp_filtered):,}개.")
    else:
        df_hp_filtered = df_hp.copy()

    # 메모리 해제
    del df_hp
    gc.collect()

    # 2. 'user_id'를 기준으로 'session_id'의 고유 개수 집계
    # user_id가 문자열일 수 있으므로 정규화합니다.
    df_hp_filtered['user_id'] = df_hp_filtered['user_id'].astype(str).str.strip().str.lower()

    print("\n⏳ `user_id`별 `session_id` 개수 집계 중...")
    sessions_per_user = df_hp_filtered.groupby('user_id')['session_id'].nunique()
    print("✅ 집계 완료.")

    # 3. 주요 통계량 확인
    print("\n📊 `user_id`별 `session_id` 개수 통계:")
    print(sessions_per_user.describe())

    # 4. 가장 많은 세션을 가진 user_id (상위 10개)
    print("\n📈 가장 많은 세션을 가진 `user_id` (상위 10개):")
    print(sessions_per_user.nlargest(10))

    # 5. 세션 개수 분포 시각화 (히스토그램)
    plt.figure(figsize=(12, 6))
    
    # 너무 많은 세션을 가진 극단값 때문에 분포가 잘 안 보일 수 있으므로,
    # 일정 개수 이상의 세션을 가진 사용자들은 하나의 빈으로 묶어서 보여줄 수 있습니다.
    # 예를 들어, 100개 이상의 세션을 가진 사용자는 '100+'로 표현.
    # 상위 99% 지점의 세션 개수를 확인하여 xlim을 설정하는 것이 좋습니다.
    q99 = sessions_per_user.quantile(0.99)
    
    sns.histplot(sessions_per_user[sessions_per_user <= q99], bins=50, kde=True)
    plt.title('Distribution of Sessions per User (Excluding Top 1%)', fontsize=16)
    plt.xlabel('Number of Sessions', fontsize=12)
    plt.ylabel('Number of Users', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # 6. 추가적으로, 1개 세션을 가진 유저의 비율, 2개 이상 세션을 가진 유저의 비율 등
    total_users = len(sessions_per_user)
    users_with_1_session = (sessions_per_user == 1).sum()
    users_with_multiple_sessions = (sessions_per_user > 1).sum()

    print(f"\n📊 세션 개수별 사용자 비율:")
    if total_users > 0:
        print(f"• 1개 세션을 가진 사용자: {users_with_1_session:,}명 ({users_with_1_session / total_users * 100:.2f}%)")
        print(f"• 2개 이상 세션을 가진 사용자: {users_with_multiple_sessions:,}명 ({users_with_multiple_sessions / total_users * 100:.2f}%)")
    else:
        print("분석할 유효한 사용자 데이터가 없습니다.")


except Exception as e:
    print(f"❌ 데이터 로드 또는 분석 중 오류 발생: {e}")
    print("파일 경로를 다시 확인하거나, 데이터에 문제가 있을 수 있습니다.")

finally:
    # 사용한 변수들 메모리에서 해제
    if 'df_hp_filtered' in locals():
        del df_hp_filtered
    if 'sessions_per_user' in locals():
        del sessions_per_user
    gc.collect()
    print("\n🗑️ 사용 변수 메모리에서 해제 완료.")
    print("=" * 80)
````

````python
import pandas as pd
import numpy as np
import gc
from datetime import timedelta

print("🔍 사용자별 첫 이벤트 시간과 마지막 이벤트 시간 데이터프레임 생성")
print("=" * 80)

# 파일 경로 정의
path_hackle_events = "gs://sprintda05_final_project/hackle/hackle_events.parquet"
path_hackle_properties = "gs://sprintda05_final_project/hackle/hackle_properties.parquet"
path_accounts_user = "gs://sprintda05_final_project/votes/accounts_user.parquet" # 유효 user_id 필터링을 위함

# 1. `hackle_events`에서 필요한 컬럼만 로드: `session_id`, `event_datetime`
print("⏳ `hackle_events`에서 `session_id`와 `event_datetime` 로드 중...")
try:
    df_events_slim = pd.read_parquet(
        path_hackle_events,
        columns=['session_id', 'event_datetime']
    )
    print(f"✅ `hackle_events` 데이터 로드 완료: 총 {len(df_events_slim):,}개 행.")
except Exception as e:
    print(f"❌ `hackle_events` 로드 중 오류 발생: {e}")
    exit() # 오류 발생 시 스크립트 종료

# 2. `hackle_properties`에서 필요한 컬럼만 로드: `session_id`, `user_id`
print("⏳ `hackle_properties`에서 `session_id`와 `user_id` 로드 중...")
try:
    df_properties_slim = pd.read_parquet(
        path_hackle_properties,
        columns=['session_id', 'user_id']
    )
    print(f"✅ `hackle_properties` 데이터 로드 완료: 총 {len(df_properties_slim):,}개 행.")
except Exception as e:
    print(f"❌ `hackle_properties` 로드 중 오류 발생: {e}")
    # df_events_slim도 여기서 해제
    del df_events_slim
    gc.collect()
    exit()

# 3. `accounts_user`에서 유효 user_id 로드 및 정규화 (이전에 확인된 유효 사용자 ID 집합)
print("⏳ `accounts_user`에서 유효 `user_id` 로드 중...")
try:
    df_au_id = pd.read_parquet(path_accounts_user, columns=['id'])
    # accounts_user.id는 int64 였으므로, hackle_properties.user_id와 동일하게 str로 정규화
    valid_user_ids = set(df_au_id['id'].astype(str).str.strip().str.lower().unique())
    print(f"✅ 유효 `user_id` {len(valid_user_ids):,}개 로드 완료.")
    del df_au_id
    gc.collect()
except Exception as e:
    print(f"❌ `accounts_user` 로드 중 오류 발생: {e}")
    # 모든 중간 df 해제
    del df_events_slim, df_properties_slim
    gc.collect()
    exit()

# 4. 'session_id' 컬럼 정규화 (조인 전 필수)
# 두 데이터프레임의 session_id를 동일한 형식으로 통일
df_events_slim['session_id'] = df_events_slim['session_id'].astype(str).str.strip().str.lower()
df_properties_slim['session_id'] = df_properties_slim['session_id'].astype(str).str.strip().str.lower()
df_properties_slim['user_id'] = df_properties_slim['user_id'].astype(str).str.strip().str.lower() # user_id도 정규화

# 5. `hackle_events`와 `hackle_properties`를 `session_id` 기준으로 조인
print("⏳ `hackle_events`와 `hackle_properties` 조인 중... (on `session_id`)")
# inner join: 양쪽에 모두 존재하는 session_id만 남김 (이미 100% 일치 확인)
df_merged = pd.merge(df_events_slim, df_properties_slim, on='session_id', how='inner')
print(f"✅ 조인 완료: 총 {len(df_merged):,}개 행.")

# 메모리 해제
del df_events_slim, df_properties_slim
gc.collect()

# 6. 유효 `user_id` 필터링
# accounts_user에 존재하지 않는 user_id는 분석에서 제외 (익명 또는 불완전한 user_id)
df_merged_filtered = df_merged[df_merged['user_id'].isin(valid_user_ids)].copy()
print(f"✅ 유효 `user_id` 필터링 완료: 총 {len(df_merged_filtered):,}개 행.")

# 메모리 해제
del df_merged
del valid_user_ids # valid_user_ids set도 더 이상 필요 없음
gc.collect()


# 7. `user_id`별 첫 이벤트 시간 (min)과 마지막 이벤트 시간 (max) 집계
print("⏳ `user_id`별 첫/마지막 이벤트 시간 집계 중...")
user_activity_summary_df = df_merged_filtered.groupby('user_id')['event_datetime'].agg(
    first_event_time='min',
    last_event_time='max'
).reset_index()
print("✅ 집계 완료.")

# 메모리 해제
del df_merged_filtered
gc.collect()

# 8. 활동 기간(duration) 컬럼 추가 (선택 사항)
user_activity_summary_df['activity_duration'] = user_activity_summary_df['last_event_time'] - user_activity_summary_df['first_event_time']

print("\n--- 사용자별 첫/마지막 이벤트 시간 데이터프레임 (상위 5개) ---")
print(user_activity_summary_df.head())

print("\n--- 사용자 활동 기간 통계 ---")
print(user_activity_summary_df['activity_duration'].describe())

# 추가 분석: 활동 기간이 0인 사용자 (단일 이벤트만 발생한 사용자)
users_single_event = (user_activity_summary_df['activity_duration'] == timedelta(seconds=0)).sum()
total_users_summary = len(user_activity_summary_df)
if total_users_summary > 0:
    print(f"\n📊 단일 이벤트만 발생한 사용자 수: {users_single_event:,}명 ({users_single_event / total_users_summary * 100:.2f}%)")
else:
    print("\n📊 분석할 유효한 사용자 활동 데이터가 없습니다.")


print("\n🗑️ 사용 변수 메모리에서 해제 완료.")
print("=" * 80)

# 최종 결과 DataFrame을 필요에 따라 저장하거나 추가 분석에 활용
# 예: user_activity_summary_df.to_parquet("user_event_timeline_summary.parquet")
````

````python
import pandas as pd
from google.cloud import storage
import os

def check_file_structure():
    """GCS 버킷의 파일 구조 확인"""
    
    bucket_name = "sprintda05_final_project"
    
    try:
        # GCS 클라이언트 생성
        client = storage.Client()
        bucket = client.bucket(bucket_name)
        
        print(f"📁 {bucket_name} 버킷 내용:")
        print("="*50)
        
        # 모든 파일 나열
        blobs = list(bucket.list_blobs())
        
        # 폴더별로 정리
        folders = {}
        for blob in blobs:
            path_parts = blob.name.split('/')
            if len(path_parts) > 1:
                folder = path_parts[0]
                if folder not in folders:
                    folders[folder] = []
                folders[folder].append(blob.name)
        
        for folder, files in folders.items():
            print(f"\n📂 {folder}/")
            for file in sorted(files):
                file_name = file.split('/')[-1]
                if file_name:  # 빈 문자열이 아닌 경우만
                    print(f"   📄 {file}")
        
        # hackle과 votes 폴더 특히 확인
        print(f"\n🔍 hackle 폴더 상세:")
        hackle_files = [f for f in blobs if 'hackle' in f.name]
        for file in hackle_files:
            print(f"   {file.name}")
            
        print(f"\n🔍 votes 폴더 상세:")
        votes_files = [f for f in blobs if 'votes' in f.name]
        for file in votes_files:
            print(f"   {file.name}")
            
    except Exception as e:
        print(f"❌ 버킷 확인 중 오류: {e}")
        
        # 대안: 직접 파일 접근 시도
        print("\n🔄 직접 파일 접근 시도...")
        
        possible_paths = [
            "gs://sprintda05_final_project/hackle/user_properties.parquet",
            "gs://sprintda05_final_project/hackle/hackle_user_properties.parquet", 
            "gs://sprintda05_final_project/hackle/user_properties/user_properties.parquet",
            "gs://sprintda05_final_project/votes/accounts_user.parquet",
            "gs://sprintda05_final_project/votes/accounts_group.parquet"
        ]
        
        for path in possible_paths:
            try:
                # 파일 존재 여부 확인
                df_test = pd.read_parquet(path, nrows=1)
                print(f"   ✅ {path} - 존재함 (컬럼: {list(df_test.columns)})")
            except Exception as e:
                print(f"   ❌ {path} - 없음")

# 파일 구조 확인 실행
check_file_structure()
````

````python
import pandas as pd
import gc
import matplotlib.pyplot as plt
import seaborn as sns

# 올바른 파일 경로 설정
path_accounts_user = "gs://sprintda05_final_project/votes/accounts_user.parquet"
path_hackle_user_properties = "gs://sprintda05_final_project/hackle/user_properties.parquet"

def analyze_user_mapping_efficient():
    """메모리 효율적인 User ID 매핑 분석"""
    
    print("=== User ID 매핑 분석 시작 ===\n")
    
    # 1. 기본 정보 로드 (필요한 컬럼만)
    print("📊 1. 기본 데이터 로드 중...")
    
    try:
        # hackle에서 user_id만 로드
        print("   - hackle user_properties 로딩...")
        hackle_df = pd.read_parquet(path_hackle_user_properties, columns=['user_id'])
        hackle_ids = hackle_df['user_id'].dropna()
        hackle_total = len(hackle_ids)
        hackle_unique_count = hackle_ids.nunique()
        
        print(f"   ✓ hackle 총 레코드: {hackle_total:,}개")
        print(f"   ✓ hackle 고유 user_id: {hackle_unique_count:,}개")
        
        # accounts에서 id만 로드
        print("   - accounts_user 로딩...")
        accounts_df = pd.read_parquet(path_accounts_user, columns=['id'])
        accounts_ids = accounts_df['id'].dropna()
        accounts_total = len(accounts_ids)
        accounts_unique_count = accounts_ids.nunique()
        
        print(f"   ✓ accounts 총 레코드: {accounts_total:,}개")
        print(f"   ✓ accounts 고유 id: {accounts_unique_count:,}개")
        
    except Exception as e:
        print(f"❌ 데이터 로드 오류: {e}")
        return None
    
    # 2. 고유값 집합으로 변환 (메모리 절약)
    print("\n🔍 2. 고유값 집합 생성 및 매핑 분석...")
    
    hackle_unique_set = set(hackle_ids.unique())
    accounts_unique_set = set(accounts_ids.unique())
    
    # 원본 DataFrame과 Series 메모리 해제
    del hackle_df, accounts_df, hackle_ids, accounts_ids
    gc.collect()
    
    # 3. 교집합 분석
    intersection = hackle_unique_set & accounts_unique_set
    only_in_hackle = hackle_unique_set - accounts_unique_set
    only_in_accounts = accounts_unique_set - hackle_unique_set
    
    # 매칭률 계산
    hackle_match_rate = (len(intersection) / len(hackle_unique_set)) * 100
    accounts_match_rate = (len(intersection) / len(accounts_unique_set)) * 100
    
    print(f"   ✓ 공통 ID (매칭됨): {len(intersection):,}개")
    print(f"   ✓ hackle에만 있음: {len(only_in_hackle):,}개")
    print(f"   ✓ accounts에만 있음: {len(only_in_accounts):,}개")
    print(f"   ✓ hackle → accounts 매칭률: {hackle_match_rate:.2f}%")
    print(f"   ✓ accounts → hackle 매칭률: {accounts_match_rate:.2f}%")
    
    # 4. 매칭되지 않는 accounts 사용자 특성 분석
    print("\n📈 3. 매칭되지 않는 accounts 사용자 특성 분석...")
    
    if len(only_in_accounts) > 0:
        try:
            # accounts 전체 데이터에서 필요한 컬럼만 로드
            accounts_analysis = pd.read_parquet(
                path_accounts_user, 
                columns=['id', 'ban_status', 'created_at', 'is_superuser', 'is_staff']
            )
            
            # 매칭되지 않는 사용자만 필터링
            unmatched_accounts = accounts_analysis[
                accounts_analysis['id'].isin(only_in_accounts)
            ].copy()
            
            print(f"   ✓ 분석 대상 (매칭안됨): {len(unmatched_accounts):,}명")
            
            # ban_status 분석
            if 'ban_status' in unmatched_accounts.columns:
                ban_status_dist = unmatched_accounts['ban_status'].value_counts()
                print("\n   📊 ban_status 분포:")
                for status, count in ban_status_dist.items():
                    percentage = (count / len(unmatched_accounts)) * 100
                    print(f"      {status}: {count:,}명 ({percentage:.1f}%)")
            
            # 생성일자 분석
            if 'created_at' in unmatched_accounts.columns:
                unmatched_accounts['created_year'] = pd.to_datetime(
                    unmatched_accounts['created_at']
                ).dt.year
                
                year_dist = unmatched_accounts['created_year'].value_counts().sort_index()
                print(f"\n   📅 계정 생성 연도 분포:")
                for year, count in year_dist.items():
                    if pd.notna(year):
                        print(f"      {int(year)}: {count:,}명")
            
            # superuser/staff 분석
            if 'is_superuser' in unmatched_accounts.columns:
                superuser_count = unmatched_accounts['is_superuser'].sum()
                staff_count = unmatched_accounts['is_staff'].sum()
                print(f"\n   👥 특수 계정:")
                print(f"      superuser: {superuser_count:,}명")
                print(f"      staff: {staff_count:,}명")
            
            del accounts_analysis, unmatched_accounts
            gc.collect()
            
        except Exception as e:
            print(f"   ⚠️ 특성 분석 중 오류: {e}")
    
    # 5. 샘플 ID 확인
    print("\n🔍 4. 샘플 ID 확인...")
    
    if len(intersection) > 0:
        sample_matched = list(intersection)[:5]
        print(f"   매칭된 ID 샘플: {sample_matched}")
    
    if len(only_in_hackle) > 0:
        sample_hackle_only = list(only_in_hackle)[:5]
        print(f"   hackle 전용 ID 샘플: {sample_hackle_only}")
    
    if len(only_in_accounts) > 0:
        sample_accounts_only = list(only_in_accounts)[:5]
        print(f"   accounts 전용 ID 샘플: {sample_accounts_only}")
    
    # 6. 요약 정보
    print(f"\n{'='*50}")
    print("📋 분석 요약")
    print(f"{'='*50}")
    print(f"🎯 매칭 성공: {len(intersection):,}명")
    print(f"📊 hackle 매칭률: {hackle_match_rate:.2f}%")
    print(f"📊 accounts 매칭률: {accounts_match_rate:.2f}%")
    print(f"🔄 분석에 사용 가능한 데이터: {len(intersection):,}명")
    
    # 결과 반환
    results = {
        'intersection': intersection,
        'only_in_hackle': only_in_hackle,
        'only_in_accounts': only_in_accounts,
        'hackle_match_rate': hackle_match_rate,
        'accounts_match_rate': accounts_match_rate,
        'hackle_total': hackle_unique_count,
        'accounts_total': accounts_unique_count
    }
    
    return results

def create_joined_sample(intersection, sample_size=1000):
    """매칭된 데이터의 샘플 조인 테이블 생성"""
    
    if len(intersection) == 0:
        print("❌ 매칭된 데이터가 없어 샘플을 생성할 수 없습니다.")
        return None
    
    print(f"\n🔗 5. 조인된 샘플 데이터 생성 (최대 {sample_size:,}개)...")
    
    try:
        # 샘플 ID 선택
        actual_sample_size = min(sample_size, len(intersection))
        sample_ids = list(intersection)[:actual_sample_size]
        
        print(f"   샘플 ID 수: {len(sample_ids):,}개")
        
        # hackle 데이터 로드 (샘플만)
        hackle_sample = pd.read_parquet(path_hackle_user_properties)
        hackle_sample = hackle_sample[hackle_sample['user_id'].isin(sample_ids)]
        print(f"   ✓ hackle 샘플: {len(hackle_sample):,}행")
        
        # accounts 데이터 로드 (샘플만)
        accounts_sample = pd.read_parquet(path_accounts_user)
        accounts_sample = accounts_sample[accounts_sample['id'].isin(sample_ids)]
        print(f"   ✓ accounts 샘플: {len(accounts_sample):,}행")
        
        # 조인 수행
        joined_sample = hackle_sample.merge(
            accounts_sample,
            left_on='user_id',
            right_on='id',
            how='inner',
            suffixes=('_hackle', '_accounts')
        )
        
        print(f"   ✅ 조인 완료: {len(joined_sample):,}행")
        print(f"   📊 조인된 컬럼 수: {len(joined_sample.columns)}개")
        
        # 조인된 데이터 기본 정보
        print(f"\n   조인된 데이터 컬럼:")
        for i, col in enumerate(joined_sample.columns):
            print(f"      {i+1:2d}. {col}")
        
        return joined_sample
        
    except Exception as e:
        print(f"   ❌ 샘플 생성 중 오류: {e}")
        return None

def visualize_mapping_results(results):
    """매핑 결과 시각화"""
    
    print(f"\n📊 6. 결과 시각화...")
    
    try:
        plt.rcParams['font.family'] = 'DejaVu Sans'  # 한글 폰트 문제 방지
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        
        # 1. 매칭 현황 파이 차트
        labels = ['Matched', 'Hackle Only', 'Accounts Only']
        sizes = [
            len(results['intersection']), 
            len(results['only_in_hackle']), 
            len(results['only_in_accounts'])
        ]
        colors = ['#2ecc71', '#3498db', '#e74c3c']
        
        ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
        ax1.set_title('User ID Mapping Status', fontsize=14, fontweight='bold')
        
        # 2. 매칭률 바 차트
        datasets = ['hackle→accounts', 'accounts→hackle']
        match_rates = [results['hackle_match_rate'], results['accounts_match_rate']]
        
        bars = ax2.bar(datasets, match_rates, color=['#3498db', '#e74c3c'])
        ax2.set_ylabel('Match Rate (%)')
        ax2.set_title('Cross-Dataset Match Rates', fontweight='bold')
        ax2.set_ylim(0, 100)
        
        # 막대 위에 수치 표시
        for bar, rate in zip(bars, match_rates):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{rate:.1f}%', ha='center', va='bottom', fontweight='bold')
        
        # 3. 전체 데이터 크기 비교
        dataset_names = ['hackle\n(unique user_id)', 'accounts\n(unique id)']
        dataset_sizes = [results['hackle_total'], results['accounts_total']]
        
        bars3 = ax3.bar(dataset_names, dataset_sizes, color=['#9b59b6', '#f39c12'])
        ax3.set_ylabel('Unique ID Count')
        ax3.set_title('Dataset Size Comparison', fontweight='bold')
        
        # 막대 위에 수치 표시
        for bar, size in zip(bars3, dataset_sizes):
            height = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2., height,
                    f'{size:,}', ha='center', va='bottom', fontweight='bold')
        
        # 4. 매칭 효율성 지표
        efficiency_labels = ['Overall\nMatch Ratio', 'Hackle\nUsability']
        efficiency_values = [
            (len(results['intersection']) / max(results['hackle_total'], results['accounts_total'])) * 100,
            results['hackle_match_rate']
        ]
        
        bars4 = ax4.bar(efficiency_labels, efficiency_values, color=['#1abc9c', '#34495e'])
        ax4.set_ylabel('Ratio (%)')
        ax4.set_title('Mapping Efficiency Metrics', fontweight='bold')
        ax4.set_ylim(0, 100)
        
        for bar, value in zip(bars4, efficiency_values):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{value:.1f}%', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.show()
        
        print("   ✅ 시각화 완료")
        
    except Exception as e:
        print(f"   ❌ 시각화 중 오류: {e}")

# 메인 실행 함수
def main():
    """메인 실행 함수"""
    
    print("🚀 User ID 매핑 분석을 시작합니다...\n")
    
    try:
        # 1. 매핑 분석 실행
        results = analyze_user_mapping_efficient()
        
        if results is None:
            print("❌ 분석 실패")
            return
        
        # 2. 샘플 조인 데이터 생성 (선택사항)
        joined_sample = create_joined_sample(results['intersection'], sample_size=1000)
        
        # 3. 시각화 (선택사항)
        visualize_mapping_results(results)
        
        # 4. 최종 권장사항
        print(f"\n{'='*50}")
        print("💡 권장사항")
        print(f"{'='*50}")
        
        if results['hackle_match_rate'] > 80:
            print("✅ 매칭률이 높음: 대부분의 hackle 사용자를 분석에 포함할 수 있습니다.")
        elif results['hackle_match_rate'] > 50:
            print("⚠️ 매칭률이 보통: 일부 hackle 사용자가 accounts에 없을 수 있습니다.")
        else:
            print("🔍 매칭률이 낮음: hackle과 accounts 데이터 간 불일치를 확인해보세요.")
        
        print(f"📊 분석 가능한 사용자: {len(results['intersection']):,}명")
        print("🔗 조인된 데이터를 활용하여 사용자 행동 분석을 진행하세요.")
        
        return results, joined_sample
        
    except Exception as e:
        print(f"❌ 메인 실행 중 오류: {e}")
        return None, None
    finally:
        # 최종 메모리 정리
        gc.collect()

# 실행
if __name__ == "__main__":
    results, sample_data = main()
````

````python
import pandas as pd
import gc
import matplotlib.pyplot as plt

# 파일 경로
path_accounts_user = "gs://sprintda05_final_project/votes/accounts_user.parquet"
path_hackle_user_properties = "gs://sprintda05_final_project/hackle/user_properties.parquet"

def analyze_user_mapping_corrected():
    """올바른 매칭 방법으로 메모리 효율적 분석"""
    
    print("🎯 올바른 매칭 방법으로 User ID 매핑 분석")
    print("="*60)
    
    # 1. 기본 데이터 로드
    print("📊 1. 데이터 로드 및 문자열 변환...")
    
    try:
        # hackle user_id 로드 (이미 문자열)
        hackle_df = pd.read_parquet(path_hackle_user_properties, columns=['user_id'])
        hackle_ids = hackle_df['user_id'].dropna()
        hackle_total = len(hackle_ids)
        hackle_unique_count = hackle_ids.nunique()
        
        print(f"   ✓ hackle 총 레코드: {hackle_total:,}개")
        print(f"   ✓ hackle 고유 user_id: {hackle_unique_count:,}개")
        
        # accounts id를 문자열로 변환하여 로드
        accounts_df = pd.read_parquet(path_accounts_user, columns=['id'])
        accounts_ids_str = accounts_df['id'].astype(str)
        accounts_total = len(accounts_ids_str)
        accounts_unique_count = accounts_ids_str.nunique()
        
        print(f"   ✓ accounts 총 레코드: {accounts_total:,}개")
        print(f"   ✓ accounts 고유 id (문자열): {accounts_unique_count:,}개")
        
    except Exception as e:
        print(f"❌ 데이터 로드 오류: {e}")
        return None
    
    # 2. 고유값 집합으로 변환
    print("\n🔍 2. 매핑 분석 (문자열 기준)...")
    
    hackle_unique_set = set(hackle_ids.unique())
    accounts_unique_set = set(accounts_ids_str.unique())
    
    # 메모리 해제
    del hackle_df, accounts_df, hackle_ids, accounts_ids_str
    gc.collect()
    
    # 3. 교집합 분석
    intersection = hackle_unique_set & accounts_unique_set
    only_in_hackle = hackle_unique_set - accounts_unique_set
    only_in_accounts = accounts_unique_set - hackle_unique_set
    
    # 매칭률 계산
    hackle_match_rate = (len(intersection) / len(hackle_unique_set)) * 100
    accounts_match_rate = (len(intersection) / len(accounts_unique_set)) * 100
    
    print(f"   ✅ 공통 ID (매칭됨): {len(intersection):,}개")
    print(f"   📊 hackle에만 있음: {len(only_in_hackle):,}개")
    print(f"   📊 accounts에만 있음: {len(only_in_accounts):,}개")
    print(f"   🎯 hackle → accounts 매칭률: {hackle_match_rate:.2f}%")
    print(f"   🎯 accounts → hackle 매칭률: {accounts_match_rate:.2f}%")
    
    # 4. 매칭되지 않는 사용자 특성 분석
    print("\n📈 3. 매칭되지 않는 사용자 특성 분석...")
    
    # hackle에만 있는 사용자 분석 (Firebase UID 등)
    if len(only_in_hackle) > 0:
        print(f"\n   🔍 hackle 전용 사용자 ({len(only_in_hackle):,}명):")
        sample_hackle_only = list(only_in_hackle)[:10]
        
        # Firebase UID 패턴 확인
        firebase_pattern = [uid for uid in sample_hackle_only if len(uid) > 10 and not uid.isdigit()]
        numeric_pattern = [uid for uid in sample_hackle_only if uid.isdigit()]
        
        print(f"      Firebase UID 형태 샘플: {firebase_pattern[:3]}")
        print(f"      숫자 형태 샘플: {numeric_pattern[:3]}")
        
        # Firebase UID 비율 추정
        firebase_count = sum(1 for uid in only_in_hackle if len(uid) > 10 and not uid.isdigit())
        firebase_ratio = (firebase_count / len(only_in_hackle)) * 100
        print(f"      Firebase UID 추정 비율: {firebase_ratio:.1f}%")
    
    # accounts에만 있는 사용자 분석
    if len(only_in_accounts) > 0:
        try:
            accounts_analysis = pd.read_parquet(
                path_accounts_user, 
                columns=['id', 'ban_status', 'created_at', 'is_superuser', 'is_staff']
            )
            
            # 문자열로 변환하여 필터링
            accounts_analysis['id_str'] = accounts_analysis['id'].astype(str)
            unmatched_accounts = accounts_analysis[
                accounts_analysis['id_str'].isin(only_in_accounts)
            ]
            
            print(f"\n   🔍 accounts 전용 사용자 ({len(unmatched_accounts):,}명):")
            
            # ban_status 분석
            if len(unmatched_accounts) > 0:
                ban_status_dist = unmatched_accounts['ban_status'].value_counts()
                print(f"      ban_status 분포:")
                for status, count in ban_status_dist.head().items():
                    percentage = (count / len(unmatched_accounts)) * 100
                    print(f"        {status}: {count:,}명 ({percentage:.1f}%)")
            
            del accounts_analysis, unmatched_accounts
            gc.collect()
            
        except Exception as e:
            print(f"   ⚠️ accounts 특성 분석 중 오류: {e}")
    
    # 5. 요약
    print(f"\n{'='*60}")
    print("📋 최종 분석 결과")
    print(f"{'='*60}")
    print(f"🎯 매칭 성공: {len(intersection):,}명")
    print(f"📊 hackle 매칭률: {hackle_match_rate:.2f}%")
    print(f"📊 accounts 매칭률: {accounts_match_rate:.2f}%")
    print(f"🔄 분석 가능한 데이터: {len(intersection):,}명")
    print(f"💡 해결책: accounts의 id를 문자열로 변환하여 매칭")
    
    results = {
        'intersection': intersection,
        'only_in_hackle': only_in_hackle,
        'only_in_accounts': only_in_accounts,
        'hackle_match_rate': hackle_match_rate,
        'accounts_match_rate': accounts_match_rate,
        'hackle_total': hackle_unique_count,
        'accounts_total': accounts_unique_count,
        'matching_method': 'string_conversion'
    }
    
    return results

def create_joined_sample_corrected(intersection, sample_size=1000):
    """올바른 매칭 방법으로 조인된 샘플 생성"""
    
    print(f"\n🔗 4. 조인된 샘플 데이터 생성 (최대 {sample_size:,}개)...")
    
    try:
        # 샘플 ID 선택
        actual_sample_size = min(sample_size, len(intersection))
        sample_ids = list(intersection)[:actual_sample_size]
        
        print(f"   샘플 ID 수: {len(sample_ids):,}개")
        
        # hackle 데이터 로드 (샘플만)
        hackle_sample = pd.read_parquet(path_hackle_user_properties)
        hackle_sample = hackle_sample[hackle_sample['user_id'].isin(sample_ids)]
        print(f"   ✓ hackle 샘플: {len(hackle_sample):,}행")
        
        # accounts 데이터 로드 (샘플만)
        accounts_sample = pd.read_parquet(path_accounts_user)
        # id를 문자열로 변환하여 매칭
        accounts_sample['id_str'] = accounts_sample['id'].astype(str)
        accounts_sample = accounts_sample[accounts_sample['id_str'].isin(sample_ids)]
        print(f"   ✓ accounts 샘플: {len(accounts_sample):,}행")
        
        # 조인 수행 (문자열 기준)
        joined_sample = hackle_sample.merge(
            accounts_sample,
            left_on='user_id',
            right_on='id_str',
            how='inner',
            suffixes=('_hackle', '_accounts')
        )
        
        # 불필요한 컬럼 제거
        if 'id_str' in joined_sample.columns:
            joined_sample = joined_sample.drop('id_str', axis=1)
        
        print(f"   ✅ 조인 완료: {len(joined_sample):,}행")
        print(f"   📊 조인된 컬럼 수: {len(joined_sample.columns)}개")
        
        # 주요 컬럼 확인
        print(f"\n   주요 컬럼:")
        key_columns = ['user_id', 'id', 'gender_hackle', 'gender_accounts', 'school_id', 'group_id', 'ban_status']
        available_columns = [col for col in key_columns if col in joined_sample.columns]
        for col in available_columns:
            print(f"      ✓ {col}")
        
        return joined_sample
        
    except Exception as e:
        print(f"   ❌ 샘플 생성 중 오류: {e}")
        return None

def visualize_corrected_results(results):
    """수정된 결과 시각화"""
    
    print(f"\n📊 5. 결과 시각화...")
    
    try:
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        
        # 1. 매칭 현황
        labels = ['Matched\n(98.05%)', 'Hackle Only\n(Firebase UID)', 'Accounts Only']
        sizes = [
            len(results['intersection']), 
            len(results['only_in_hackle']), 
            len(results['only_in_accounts'])
        ]
        colors = ['#2ecc71', '#3498db', '#e74c3c']
        
        ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
        ax1.set_title('User ID Mapping Results\n(String Conversion)', fontweight='bold')
        
        # 2. 매칭률 비교
        categories = ['hackle→accounts', 'accounts→hackle']
        rates = [results['hackle_match_rate'], results['accounts_match_rate']]
        
        bars = ax2.bar(categories, rates, color=['#3498db', '#e74c3c'])
        ax2.set_ylabel('Match Rate (%)')
        ax2.set_title('Cross-Dataset Match Rates', fontweight='bold')
        ax2.set_ylim(0, 100)
        
        for bar, rate in zip(bars, rates):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{rate:.1f}%', ha='center', va='bottom', fontweight='bold')
        
        # 3. 데이터 가용성
        usable_data = len(results['intersection'])
        total_hackle = results['hackle_total']
        
        ax3.bar(['Available for Analysis'], [usable_data], color='#2ecc71', width=0.5)
        ax3.bar(['Total hackle Users'], [total_hackle], color='#95a5a6', alpha=0.7, width=0.3)
        ax3.set_ylabel('User Count')
        ax3.set_title('Data Availability for Analysis', fontweight='bold')
        ax3.text(0, usable_data + 5000, f'{usable_data:,}', ha='center', fontweight='bold')
        
        # 4. 매칭 성공률 히스토리
        methods = ['Original\n(Type Mismatch)', 'String\nConversion']
        success_rates = [0, results['hackle_match_rate']]
        
        bars4 = ax4.bar(methods, success_rates, color=['#e74c3c', '#2ecc71'])
        ax4.set_ylabel('Success Rate (%)')
        ax4.set_title('Matching Method Comparison', fontweight='bold')
        ax4.set_ylim(0, 100)
        
        for bar, rate in zip(bars4, success_rates):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{rate:.1f}%', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.show()
        
        print("   ✅ 시각화 완료")
        
    except Exception as e:
        print(f"   ❌ 시각화 중 오류: {e}")

# 수정된 분석 실행
print("🚀 올바른 매칭 방법으로 전체 분석 시작...")

try:
    # 1. 수정된 매핑 분석
    corrected_results = analyze_user_mapping_corrected()
    
    if corrected_results:
        # 2. 조인된 샘플 생성
        joined_sample = create_joined_sample_corrected(corrected_results['intersection'], 1000)
        
        # 3. 시각화
        visualize_corrected_results(corrected_results)
        
        # 4. 최종 권장사항
        print(f"\n{'='*60}")
        print("💡 최종 권장사항 및 다음 단계")
        print(f"{'='*60}")
        print("✅ 매칭 문제 해결됨: accounts의 id를 문자열로 변환")
        print(f"📊 98.05%의 높은 매칭률 달성")
        print(f"🔄 {len(corrected_results['intersection']):,}명의 사용자 데이터 분석 가능")
        print(f"📋 다음 분석 가능 항목:")
        print("   - 사용자 행동 패턴 분석 (hackle events + accounts info)")
        print("   - 학교/학급별 사용자 특성 분석")
        print("   - 성별/학년별 앱 사용 패턴")
        print("   - ban_status별 사용자 행동 차이")
        print("   - 포인트/친구수와 앱 사용량 상관관계")
        
        print(f"\n🔧 코드에서 사용할 조인 방법:")
        print("```python")
        print("# accounts id를 문자열로 변환하여 조인")
        print("accounts_df['id_str'] = accounts_df['id'].astype(str)")
        print("joined_df = hackle_df.merge(accounts_df, left_on='user_id', right_on='id_str')")
        print("```")
        
        print(f"\n📈 분석 준비 완료! 이제 구체적인 분석을 시작할 수 있습니다.")
        
    else:
        print("❌ 분석 실패")
        
except Exception as e:
    print(f"❌ 전체 분석 중 오류: {e}")
    
finally:
    gc.collect()
````

````python
import pandas as pd
import gc

# 파일 경로들
paths = {
    'events': "gs://sprintda05_final_project/hackle/hackle_events.parquet",
    'users': "gs://sprintda05_final_project/hackle/user_properties.parquet", 
    'properties': "gs://sprintda05_final_project/hackle/hackle_properties.parquet",
    'devices': "gs://sprintda05_final_project/hackle/device_properties.parquet"
}

def check_table_columns():
    """각 테이블의 컬럼만 확인 (최소 메모리)"""
    
    print("🔍 테이블 컬럼 구조 확인")
    print("="*40)
    
    table_columns = {}
    
    for name, path in paths.items():
        try:
            # 단 1개 행만 로드하여 컬럼 확인
            df = pd.read_parquet(path).head(1)
            columns = list(df.columns)
            table_columns[name] = columns
            print(f"✅ {name}: {columns}")
            
            # 즉시 삭제
            del df
            gc.collect()
            
        except Exception as e:
            print(f"❌ {name}: {e}")
    
    return table_columns

def test_direct_id_connection():
    """events.id와 users.user_id 직접 연결 테스트"""
    
    print(f"\n🎯 직접 ID 연결 테스트...")
    
    try:
        # events id만 소량 로드
        print("   events id 로딩...")
        events_df = pd.read_parquet(paths['events'], columns=['id']).head(50)
        events_ids = set(events_df['id'].astype(str).tolist())
        del events_df
        gc.collect()
        
        print(f"   events 샘플 id: {list(events_ids)[:3]}")
        
        # users user_id만 소량 로드  
        print("   users user_id 로딩...")
        users_df = pd.read_parquet(paths['users'], columns=['user_id']).head(50)
        users_ids = set(users_df['user_id'].astype(str).tolist())
        del users_df
        gc.collect()
        
        print(f"   users 샘플 user_id: {list(users_ids)[:3]}")
        
        # 교집합 확인
        intersection = events_ids & users_ids
        print(f"   샘플 매칭: {len(intersection)}개")
        
        if len(intersection) > 0:
            print(f"   ✅ 직접 연결 가능! 매칭 샘플: {list(intersection)[:2]}")
            return True
        else:
            print(f"   ❌ 직접 연결 불가")
            return False
            
    except Exception as e:
        print(f"   ❌ 테스트 오류: {e}")
        return False

def test_session_connection():
    """session_id 간접 연결 테스트"""
    
    print(f"\n🔗 session_id 간접 연결 테스트...")
    
    try:
        # events session_id 확인
        print("   events session_id 확인...")
        events_df = pd.read_parquet(paths['events'], columns=['session_id']).head(30)
        events_sessions = set(events_df['session_id'].dropna().astype(str).tolist())
        del events_df
        gc.collect()
        
        if not events_sessions:
            print("   ❌ events에 session_id 없음")
            return False
            
        print(f"   events 샘플 session: {list(events_sessions)[:2]}")
        
        # properties session_id 확인
        print("   properties session_id 확인...")
        props_df = pd.read_parquet(paths['properties'], columns=['session_id']).head(30)
        props_sessions = set(props_df['session_id'].dropna().astype(str).tolist())
        del props_df
        gc.collect()
        
        print(f"   properties 샘플 session: {list(props_sessions)[:2]}")
        
        # 교집합 확인
        session_match = events_sessions & props_sessions
        print(f"   session 매칭: {len(session_match)}개")
        
        if len(session_match) > 0:
            print(f"   ✅ session 연결 가능!")
            
            # properties user_id 확인
            print("   properties user_id 확인...")
            props_users_df = pd.read_parquet(paths['properties'], columns=['user_id']).head(30)
            props_user_ids = set(props_users_df['user_id'].dropna().astype(str).tolist())
            del props_users_df
            gc.collect()
            
            if props_user_ids:
                print(f"   properties 샘플 user_id: {list(props_user_ids)[:2]}")
                return True
            else:
                print(f"   ❌ properties에 user_id 없음")
                return False
        else:
            print(f"   ❌ session 연결 불가")
            return False
            
    except Exception as e:
        print(f"   ❌ 테스트 오류: {e}")
        return False

def test_small_join():
    """아주 작은 샘플로 조인 테스트"""
    
    print(f"\n🧪 소규모 조인 테스트...")
    
    try:
        # 매우 작은 샘플로 조인 시도
        print("   소규모 데이터 로딩...")
        
        events_small = pd.read_parquet(paths['events']).head(20)
        print(f"   events: {len(events_small)}행")
        
        users_small = pd.read_parquet(paths['users']).head(20)  
        print(f"   users: {len(users_small)}행")
        
        # events.id를 user_id로 변경하여 조인 시도
        if 'id' in events_small.columns:
            events_small = events_small.rename(columns={'id': 'user_id'})
            
            joined = events_small.merge(users_small, on='user_id', how='inner')
            print(f"   조인 결과: {len(joined)}행")
            
            if len(joined) > 0:
                print(f"   ✅ 조인 성공!")
                print(f"   조인 컬럼: {list(joined.columns)[:5]}...")
                return True
            else:
                print(f"   ❌ 조인 실패 (매칭 없음)")
                return False
        else:
            print(f"   ❌ events에 id 컬럼 없음")
            return False
            
    except Exception as e:
        print(f"   ❌ 조인 테스트 오류: {e}")
        return False
    finally:
        # 메모리 정리
        try:
            del events_small, users_small
            if 'joined' in locals():
                del joined
        except:
            pass
        gc.collect()

def get_connection_solution():
    """연결 솔루션 제공"""
    
    print(f"\n{'='*40}")
    print("💡 연결 솔루션")
    print(f"{'='*40}")
    
    # 1. 컬럼 구조 확인
    columns = check_table_columns()
    
    # 2. 직접 연결 테스트
    direct_possible = test_direct_id_connection()
    
    # 3. 간접 연결 테스트
    indirect_possible = test_session_connection() if not direct_possible else False
    
    # 4. 조인 테스트
    join_works = test_small_join() if direct_possible else False
    
    # 5. 결론 및 코드 제안
    print(f"\n📋 분석 결과:")
    
    if direct_possible and join_works:
        print("✅ 해결책: events.id = user_properties.user_id")
        print("🔧 추천 코드:")
        print("```python")
        print("events = pd.read_parquet(path_events)")
        print("events['user_id'] = events['id']  # id를 user_id로 사용")
        print("users = pd.read_parquet(path_users)")
        print("joined = events.merge(users, on='user_id')")
        print("```")
        return "direct_connection"
        
    elif indirect_possible:
        print("✅ 해결책: session_id 간접 연결")
        print("🔧 추천 코드:")
        print("```python")
        print("events = pd.read_parquet(path_events)")
        print("props = pd.read_parquet(path_properties)")
        print("users = pd.read_parquet(path_users)")
        print("step1 = events.merge(props, on='session_id')")
        print("final = step1.merge(users, on='user_id')")
        print("```")
        return "indirect_connection"
        
    else:
        print("❌ 연결 방법을 찾지 못했습니다")
        print("💡 대안:")
        print("   - 각 테이블을 별도로 분석")
        print("   - 다른 공통 키 확인")
        print("   - 시간 기반 분석")
        return "no_connection"

# 메인 실행 (최소 메모리)
def main_minimal():
    print("🚀 최소 메모리 hackle 연결 분석")
    
    try:
        solution = get_connection_solution()
        
        print(f"\n✅ 분석 완료: {solution}")
        return solution
        
    except Exception as e:
        print(f"❌ 분석 오류: {e}")
        return None
    finally:
        gc.collect()
        print("🧹 메모리 정리 완료")

# 실행
if __name__ == "__main__":
    result = main_minimal()
````

````python
import pandas as pd
import gc

# 파일 경로
path_hackle_events = "gs://sprintda05_final_project/hackle/hackle_events.parquet"
path_hackle_user_properties = "gs://sprintda05_final_project/hackle/user_properties.parquet"
path_hackle_properties = "gs://sprintda05_final_project/hackle/hackle_properties.parquet"

def verify_user_id_connection():
    """hackle_user_properties와 hackle_properties 간 user_id 연결 확인"""
    
    print("🔍 hackle 테이블 user_id 연결 확인")
    print("="*40)
    
    try:
        # 1. user_properties의 user_id 샘플 확인
        print("📊 1. user_properties user_id 확인...")
        users_df = pd.read_parquet(path_hackle_user_properties, columns=['user_id']).head(100)
        users_sample = set(users_df['user_id'].astype(str).tolist())
        users_total = len(users_df)
        del users_df
        gc.collect()
        
        print(f"   샘플 user_id: {list(users_sample)[:3]}")
        
        # 2. hackle_properties의 user_id 샘플 확인
        print("📊 2. hackle_properties user_id 확인...")
        props_df = pd.read_parquet(path_hackle_properties, columns=['user_id']).head(100)
        props_sample = set(props_df['user_id'].astype(str).tolist())
        props_total = len(props_df)
        del props_df
        gc.collect()
        
        print(f"   샘플 user_id: {list(props_sample)[:3]}")
        
        # 3. 샘플 교집합 확인
        sample_intersection = users_sample & props_sample
        print(f"   샘플 매칭: {len(sample_intersection)}개")
        
        if len(sample_intersection) > 0:
            print(f"   ✅ user_id 연결 확인! 매칭 샘플: {list(sample_intersection)[:2]}")
            return True
        else:
            print(f"   ❌ user_id 연결 불가")
            return False
            
    except Exception as e:
        print(f"❌ 연결 확인 오류: {e}")
        return False

def verify_full_connection():
    """전체 데이터에서 user_id 연결 확인"""
    
    print(f"\n🎯 전체 데이터 user_id 연결 확인...")
    
    try:
        # user_properties 전체 user_id (메모리 효율적)
        print("   user_properties 전체 로딩...")
        users_ids = pd.read_parquet(path_hackle_user_properties, columns=['user_id'])['user_id']
        users_unique = set(users_ids.astype(str).unique())
        users_count = len(users_unique)
        del users_ids
        gc.collect()
        
        print(f"   user_properties 고유 user_id: {users_count:,}개")
        
        # hackle_properties 전체 user_id
        print("   hackle_properties 전체 로딩...")
        props_ids = pd.read_parquet(path_hackle_properties, columns=['user_id'])['user_id']
        props_unique = set(props_ids.astype(str).unique())
        props_count = len(props_unique)
        del props_ids
        gc.collect()
        
        print(f"   hackle_properties 고유 user_id: {props_count:,}개")
        
        # 교집합 계산
        intersection = users_unique & props_unique
        only_users = users_unique - props_unique
        only_props = props_unique - users_unique
        
        print(f"\n   🔗 연결 결과:")
        print(f"   공통 user_id: {len(intersection):,}개")
        print(f"   user_properties만: {len(only_users):,}개")
        print(f"   hackle_properties만: {len(only_props):,}개")
        
        # 매칭률 계산
        users_match_rate = (len(intersection) / users_count) * 100
        props_match_rate = (len(intersection) / props_count) * 100
        
        print(f"   user_properties → hackle_properties: {users_match_rate:.2f}%")
        print(f"   hackle_properties → user_properties: {props_match_rate:.2f}%")
        
        return {
            'intersection': len(intersection),
            'users_total': users_count,
            'props_total': props_count,
            'users_match_rate': users_match_rate,
            'props_match_rate': props_match_rate
        }
        
    except Exception as e:
        print(f"❌ 전체 연결 확인 오류: {e}")
        return None

def test_3way_join_path():
    """events → properties → user_properties 3-way 조인 경로 테스트"""
    
    print(f"\n🔗 3-way 조인 경로 테스트...")
    
    try:
        # 1. events와 properties의 session_id 연결 확인
        print("   1) events ↔ properties (session_id)...")
        
        events_sessions = pd.read_parquet(path_hackle_events, columns=['session_id']).head(50)
        events_sample_sessions = set(events_sessions['session_id'].astype(str).tolist())
        del events_sessions
        gc.collect()
        
        props_sessions = pd.read_parquet(path_hackle_properties, columns=['session_id']).head(50)
        props_sample_sessions = set(props_sessions['session_id'].astype(str).tolist())
        del props_sessions
        gc.collect()
        
        session_match = events_sample_sessions & props_sample_sessions
        print(f"      샘플 session 매칭: {len(session_match)}개")
        
        if len(session_match) == 0:
            print("   ❌ session_id 연결 불가")
            return False
        
        # 2. 작은 샘플로 3-way 조인 테스트
        print("   2) 3-way 조인 테스트...")
        
        events_small = pd.read_parquet(path_hackle_events).head(20)
        props_small = pd.read_parquet(path_hackle_properties).head(20)
        users_small = pd.read_parquet(path_hackle_user_properties).head(20)
        
        # 단계별 조인
        step1 = events_small.merge(props_small, on='session_id', how='inner')
        print(f"      events + properties: {len(step1)}행")
        
        if len(step1) > 0:
            step2 = step1.merge(users_small, on='user_id', how='inner')
            print(f"      최종 3-way 조인: {len(step2)}행")
            
            if len(step2) > 0:
                print(f"   ✅ 3-way 조인 성공!")
                print(f"      최종 컬럼 수: {len(step2.columns)}개")
                
                # 주요 컬럼들 확인
                key_cols = ['session_id', 'user_id', 'event_key', 'gender', 'grade', 'school_id']
                available_cols = [col for col in key_cols if col in step2.columns]
                print(f"      주요 컬럼: {available_cols}")
                
                return True
            else:
                print(f"   ❌ user_id 조인 실패")
                return False
        else:
            print(f"   ❌ session_id 조인 실패") 
            return False
            
    except Exception as e:
        print(f"❌ 3-way 조인 테스트 오류: {e}")
        return False
    finally:
        # 메모리 정리
        try:
            del events_small, props_small, users_small
            if 'step1' in locals():
                del step1
            if 'step2' in locals():
                del step2
        except:
            pass
        gc.collect()

def create_join_guide(user_connection_result, threeway_possible):
    """조인 가이드 생성"""
    
    print(f"\n{'='*50}")
    print("💡 hackle 테이블 조인 가이드")
    print(f"{'='*50}")
    
    if user_connection_result and threeway_possible:
        print("✅ 최적 해결책: 3-way 조인 가능")
        print(f"📊 user_id 매칭률: {user_connection_result['users_match_rate']:.1f}%")
        print(f"🎯 분석 가능 사용자: {user_connection_result['intersection']:,}명")
        
        print(f"\n🔧 추천 조인 코드:")
        print("```python")
        print("# 메모리 효율적 3-way 조인")
        print("events = pd.read_parquet(path_events)")
        print("props = pd.read_parquet(path_properties)")
        print("users = pd.read_parquet(path_user_properties)")
        print("")
        print("# 단계별 조인")
        print("step1 = events.merge(props, on='session_id', how='inner')")
        print("final = step1.merge(users, on='user_id', how='inner')")
        print("")
        print("# 최종 분석 데이터 준비 완료!")
        print("```")
        
        print(f"\n📊 분석 가능한 항목:")
        print("   - 사용자별 이벤트 패턴 (성별, 학년, 학교별)")
        print("   - 디바이스/OS별 사용 패턴")
        print("   - 세션별 사용자 행동 분석")
        print("   - 학교/학급별 앱 사용 패턴")
        
    elif user_connection_result:
        print("✅ 부분 해결책: user_properties ↔ hackle_properties 조인 가능")
        print(f"📊 user_id 매칭률: {user_connection_result['users_match_rate']:.1f}%")
        
        print(f"\n🔧 대안 조인 코드:")
        print("```python")
        print("props = pd.read_parquet(path_properties)")
        print("users = pd.read_parquet(path_user_properties)")
        print("joined = props.merge(users, on='user_id', how='inner')")
        print("```")
        
    else:
        print("❌ 직접 조인 불가")
        print("💡 대안: 각 테이블을 별도로 분석")

# 메인 실행
def main_verify_connection():
    print("🚀 hackle user_id 연결 검증 시작")
    
    try:
        # 1. 샘플 연결 확인
        sample_connection = verify_user_id_connection()
        
        if sample_connection:
            # 2. 전체 데이터 연결 확인
            full_connection = verify_full_connection()
            
            # 3. 3-way 조인 경로 테스트
            threeway_join = test_3way_join_path()
            
            # 4. 조인 가이드 생성
            create_join_guide(full_connection, threeway_join)
            
            return full_connection, threeway_join
        else:
            print("❌ 기본 연결 확인 실패")
            return None, False
            
    except Exception as e:
        print(f"❌ 검증 오류: {e}")
        return None, False
    finally:
        gc.collect()
        print("\n🧹 메모리 정리 완료")

# 실행
if __name__ == "__main__":
    connection_result, threeway_result = main_verify_connection()
````

````python
import pandas as pd
import gc

# 파일 경로들
paths = {
    'events': "gs://sprintda05_final_project/hackle/hackle_events.parquet",
    'users': "gs://sprintda05_final_project/hackle/user_properties.parquet", 
    'properties': "gs://sprintda05_final_project/hackle/hackle_properties.parquet"
}

def analyze_id_formats():
    """ID 형식과 데이터 타입 상세 분석"""
    
    print("🔍 ID 형식 상세 분석")
    print("="*40)
    
    try:
        # 1. events.id 분석
        print("📊 1. events.id 분석...")
        events_df = pd.read_parquet(paths['events'], columns=['id']).head(100)
        events_ids = events_df['id'].dropna()
        
        print(f"   데이터 타입: {events_ids.dtype}")
        print(f"   샘플 값: {events_ids.head(3).tolist()}")
        print(f"   길이 범위: {events_ids.str.len().min()} ~ {events_ids.str.len().max()}")
        print(f"   UUID 형태: {'✅' if '-' in str(events_ids.iloc[0]) else '❌'}")
        
        del events_df, events_ids
        gc.collect()
        
        # 2. users.user_id 분석
        print("\n📊 2. users.user_id 분석...")
        users_df = pd.read_parquet(paths['users'], columns=['user_id']).head(100)
        users_ids = users_df['user_id'].dropna()
        
        print(f"   데이터 타입: {users_ids.dtype}")
        print(f"   샘플 값: {users_ids.head(3).tolist()}")
        print(f"   숫자 형태: {'✅' if users_ids.astype(str).str.isdigit().all() else '❌'}")
        
        # 숫자 범위 확인
        try:
            numeric_ids = pd.to_numeric(users_ids, errors='coerce')
            print(f"   숫자 범위: {int(numeric_ids.min())} ~ {int(numeric_ids.max())}")
        except:
            print(f"   숫자 변환 불가")
        
        del users_df, users_ids
        gc.collect()
        
        # 3. properties.user_id 분석
        print("\n📊 3. properties.user_id 분석...")
        props_df = pd.read_parquet(paths['properties'], columns=['user_id']).head(100)
        props_ids = props_df['user_id'].fillna('')  # NaN을 빈 문자열로
        
        print(f"   데이터 타입: {props_ids.dtype}")
        print(f"   총 값: {len(props_ids)}개")
        print(f"   빈 값: {(props_ids == '').sum()}개")
        print(f"   유효 값: {(props_ids != '').sum()}개")
        
        # 유효한 값들만 분석
        valid_props = props_ids[props_ids != '']
        if len(valid_props) > 0:
            print(f"   유효 샘플: {valid_props.head(3).tolist()}")
            try:
                numeric_props = pd.to_numeric(valid_props, errors='coerce')
                numeric_props = numeric_props.dropna()
                if len(numeric_props) > 0:
                    print(f"   숫자 범위: {int(numeric_props.min())} ~ {int(numeric_props.max())}")
            except:
                print(f"   숫자 변환 불가")
        
        del props_df, props_ids
        gc.collect()
        
    except Exception as e:
        print(f"❌ ID 분석 오류: {e}")

def verify_session_connection():
    """session_id 연결 상세 검증"""
    
    print(f"\n🔗 session_id 연결 상세 검증...")
    
    try:
        # events session_id 분석
        events_df = pd.read_parquet(paths['events'], columns=['session_id']).head(200)
        events_sessions = events_df['session_id'].dropna()
        events_unique = set(events_sessions.unique())
        
        print(f"   events session 샘플: {len(events_unique)}개 고유값")
        print(f"   events session 예시: {list(events_unique)[:2]}")
        
        del events_df
        gc.collect()
        
        # properties session_id 분석
        props_df = pd.read_parquet(paths['properties'], columns=['session_id']).head(200)
        props_sessions = props_df['session_id'].dropna()
        props_unique = set(props_sessions.unique())
        
        print(f"   properties session 샘플: {len(props_unique)}개 고유값")
        print(f"   properties session 예시: {list(props_unique)[:2]}")
        
        del props_df
        gc.collect()
        
        # 교집합 확인
        session_intersection = events_unique & props_unique
        print(f"   session 매칭: {len(session_intersection)}개")
        
        if len(session_intersection) > 0:
            print(f"   ✅ session_id로 연결 가능!")
            return True
        else:
            print(f"   ❌ session_id 연결 불가")
            return False
            
    except Exception as e:
        print(f"❌ session 검증 오류: {e}")
        return False

def test_actual_join():
    """실제 조인 가능성 테스트"""
    
    print(f"\n🧪 실제 조인 테스트...")
    
    try:
        # 작은 샘플로 실제 조인 시도
        events_small = pd.read_parquet(paths['events']).head(50)
        props_small = pd.read_parquet(paths['properties']).head(50)
        users_small = pd.read_parquet(paths['users']).head(50)
        
        print(f"   로드 완료: events({len(events_small)}), props({len(props_small)}), users({len(users_small)})")
        
        # Step 1: events + properties (session_id)
        step1 = events_small.merge(props_small, on='session_id', how='inner')
        print(f"   Step 1 (events + props): {len(step1)}행")
        
        if len(step1) > 0:
            # properties의 user_id 정리 (빈 값 제거)
            step1_clean = step1[step1['user_id'] != ''].copy()
            print(f"   Step 1 정리 후: {len(step1_clean)}행")
            
            if len(step1_clean) > 0:
                # Step 2: + users (user_id)
                # user_id 타입 맞춤
                step1_clean['user_id'] = step1_clean['user_id'].astype(str)
                users_small['user_id'] = users_small['user_id'].astype(str)
                
                step2 = step1_clean.merge(users_small, on='user_id', how='inner')
                print(f"   Step 2 (+ users): {len(step2)}행")
                
                if len(step2) > 0:
                    print(f"   ✅ 전체 조인 성공!")
                    print(f"   최종 컬럼: {list(step2.columns)}")
                    
                    # 샘플 데이터 확인
                    key_cols = ['session_id', 'user_id', 'event_key', 'gender', 'grade']
                    available_cols = [col for col in key_cols if col in step2.columns]
                    print(f"   주요 데이터: {available_cols}")
                    
                    return True, len(step2)
                else:
                    print(f"   ❌ user_id 조인 실패")
                    return False, 0
            else:
                print(f"   ❌ properties에 유효한 user_id 없음")
                return False, 0
        else:
            print(f"   ❌ session_id 조인 실패")
            return False, 0
            
    except Exception as e:
        print(f"❌ 조인 테스트 오류: {e}")
        return False, 0
    finally:
        # 메모리 정리
        try:
            del events_small, props_small, users_small
            if 'step1' in locals():
                del step1
            if 'step1_clean' in locals():
                del step1_clean
            if 'step2' in locals():
                del step2
        except:
            pass
        gc.collect()

def final_solution():
    """최종 해결책 제시"""
    
    print(f"\n{'='*50}")
    print("💡 최종 분석 결과")
    print(f"{'='*50}")
    
    # 전체 분석 실행
    analyze_id_formats()
    session_ok = verify_session_connection()
    join_ok, join_count = test_actual_join()
    
    print(f"\n📋 결론:")
    
    if join_ok:
        print("✅ 해결책 확정: session_id 기반 3-way 조인")
        print(f"🎯 조인 가능한 데이터: {join_count}개 (샘플 기준)")
        print(f"\n🔧 최종 추천 코드:")
        print("```python")
        print("# 메모리 효율적 조인")
        print("events = pd.read_parquet(path_events)")
        print("props = pd.read_parquet(path_properties)")  
        print("users = pd.read_parquet(path_users)")
        print("")
        print("# 데이터 정리")
        print("props = props[props['user_id'] != '']  # 빈 user_id 제거")
        print("props['user_id'] = props['user_id'].astype(str)")
        print("users['user_id'] = users['user_id'].astype(str)")
        print("")
        print("# 조인")
        print("step1 = events.merge(props, on='session_id', how='inner')")
        print("final = step1.merge(users, on='user_id', how='inner')")
        print("```")
        
        print(f"\n⚠️ 주의사항:")
        print("   - properties의 user_id에 빈 값이 있음 → 필터링 필요")
        print("   - events.id는 UUID, users.user_id는 숫자 → 직접 연결 불가")
        print("   - session_id가 유일한 연결 고리")
        
    elif session_ok:
        print("⚠️ 부분 해결: session_id 연결 가능하나 user_id 매칭 문제")
        print("💡 대안: events + properties만 조인하여 분석")
        
    else:
        print("❌ 연결 불가: 각 테이블을 별도로 분석해야 함")
        
    return join_ok

# 실행
if __name__ == "__main__":
    result = final_solution()
````

````python
import pandas as pd
import gc
import re

# 파일 경로
path_properties = "gs://sprintda05_final_project/hackle/hackle_properties.parquet"
path_users = "gs://sprintda05_final_project/hackle/user_properties.parquet"

def analyze_properties_user_id_types():
    """hackle_properties의 user_id 타입별 상세 분석"""
    
    print("🔍 hackle_properties user_id 타입 분석")
    print("="*50)
    
    try:
        # properties user_id 전체 로드
        print("📊 전체 데이터 로딩...")
        props_df = pd.read_parquet(path_properties, columns=['user_id'])
        
        # 기본 통계
        total_count = len(props_df)
        null_count = props_df['user_id'].isnull().sum()
        empty_count = (props_df['user_id'] == '').sum()
        valid_count = total_count - null_count - empty_count
        
        print(f"   전체 레코드: {total_count:,}개")
        print(f"   NULL 값: {null_count:,}개")
        print(f"   빈 문자열: {empty_count:,}개")
        print(f"   유효 값: {valid_count:,}개")
        
        # 유효한 user_id만 분석
        valid_ids = props_df[(props_df['user_id'].notna()) & (props_df['user_id'] != '')]['user_id'].astype(str)
        unique_ids = valid_ids.unique()
        
        print(f"   고유 user_id: {len(unique_ids):,}개")
        
        del props_df
        gc.collect()
        
        # 숫자 vs 문자열 분류
        print(f"\n🔢 user_id 타입 분류...")
        
        numeric_ids = []
        string_ids = []
        mixed_ids = []
        
        for uid in unique_ids:
            uid_str = str(uid)
            if uid_str.isdigit():
                numeric_ids.append(uid_str)
            elif re.match(r'^\d+$', uid_str):  # 순수 숫자
                numeric_ids.append(uid_str)
            elif any(c.isdigit() for c in uid_str) and any(c.isalpha() for c in uid_str):
                mixed_ids.append(uid_str)  # 숫자+문자 혼합
            else:
                string_ids.append(uid_str)  # 순수 문자
        
        print(f"   순수 숫자 ID: {len(numeric_ids):,}개")
        print(f"   순수 문자 ID: {len(string_ids):,}개") 
        print(f"   혼합 ID: {len(mixed_ids):,}개")
        
        # 각 타입별 샘플 출력
        print(f"\n📋 타입별 샘플:")
        
        if numeric_ids:
            numeric_sample = sorted(numeric_ids)[:5]
            print(f"   숫자 ID 샘플: {numeric_sample}")
            
            # 숫자 범위 확인
            numeric_values = [int(x) for x in numeric_ids]
            print(f"   숫자 ID 범위: {min(numeric_values):,} ~ {max(numeric_values):,}")
        
        if string_ids:
            string_sample = sorted(string_ids)[:5]
            print(f"   문자 ID 샘플: {string_sample}")
            
            # 문자 ID 길이 분석
            string_lengths = [len(x) for x in string_ids]
            print(f"   문자 ID 길이: {min(string_lengths)} ~ {max(string_lengths)}")
        
        if mixed_ids:
            mixed_sample = sorted(mixed_ids)[:5]
            print(f"   혼합 ID 샘플: {mixed_sample}")
        
        return {
            'numeric_count': len(numeric_ids),
            'string_count': len(string_ids),
            'mixed_count': len(mixed_ids),
            'numeric_ids': numeric_ids,
            'string_ids': string_ids
        }
        
    except Exception as e:
        print(f"❌ 분석 오류: {e}")
        return None

def check_numeric_id_overlap():
    """숫자 형태의 user_id만으로 users와 매칭 확인"""
    
    print(f"\n🎯 숫자 ID만으로 매칭 테스트...")
    
    try:
        # users의 user_id
        users_df = pd.read_parquet(path_users, columns=['user_id'])
        users_numeric = set(users_df['user_id'].astype(str).tolist())
        print(f"   users 숫자 ID: {len(users_numeric):,}개")
        
        del users_df
        gc.collect()
        
        # properties 분석 결과 사용
        analysis_result = analyze_properties_user_id_types()
        
        if analysis_result and analysis_result['numeric_count'] > 0:
            props_numeric = set(analysis_result['numeric_ids'])
            print(f"   properties 숫자 ID: {len(props_numeric):,}개")
            
            # 교집합 확인
            intersection = users_numeric & props_numeric
            only_users = users_numeric - props_numeric
            only_props = props_numeric - users_numeric
            
            print(f"\n   📊 숫자 ID 매칭 결과:")
            print(f"   공통 ID: {len(intersection):,}개")
            print(f"   users만: {len(only_users):,}개")
            print(f"   properties만: {len(only_props):,}개")
            
            if len(intersection) > 0:
                match_rate = len(intersection) / len(users_numeric) * 100
                print(f"   매칭률: {match_rate:.1f}%")
                
                # 샘플 확인
                sample_matches = sorted(list(intersection))[:5]
                print(f"   매칭 샘플: {sample_matches}")
                
                return True, len(intersection)
            else:
                print(f"   ❌ 숫자 ID로도 매칭 불가")
                return False, 0
        else:
            print(f"   ❌ properties에 숫자 ID 없음")
            return False, 0
            
    except Exception as e:
        print(f"❌ 매칭 테스트 오류: {e}")
        return False, 0

def create_filtered_join_strategy():
    """필터링된 조인 전략 제시"""
    
    print(f"\n{'='*50}")
    print("💡 최적화된 조인 전략")
    print(f"{'='*50}")
    
    # 분석 실행
    analysis_result = analyze_properties_user_id_types()
    matching_works, match_count = check_numeric_id_overlap()
    
    if matching_works:
        print("✅ 최적 해결책: 숫자 user_id만 필터링하여 조인")
        print(f"🎯 조인 가능한 사용자: {match_count:,}명")
        
        print(f"\n🔧 추천 최적화 코드:")
        print("```python")
        print("import re")
        print("")
        print("# 데이터 로드")
        print("events = pd.read_parquet(path_events)")
        print("props = pd.read_parquet(path_properties)")
        print("users = pd.read_parquet(path_users)")
        print("")
        print("# properties에서 숫자 user_id만 필터링")
        print("props = props[props['user_id'] != '']  # 빈 값 제거")
        print("props['user_id_str'] = props['user_id'].astype(str)")
        print("props_numeric = props[props['user_id_str'].str.isdigit()].copy()")
        print("props_numeric['user_id'] = props_numeric['user_id_str']")
        print("")
        print("# users ID를 문자열로 변환")
        print("users['user_id'] = users['user_id'].astype(str)")
        print("")
        print("# 단계별 조인")
        print("step1 = events.merge(props_numeric, on='session_id', how='inner')")
        print("final = step1.merge(users, on='user_id', how='inner')")
        print("")
        print(f"# 결과: 약 {match_count:,}명의 사용자 데이터로 분석 가능")
        print("```")
        
        print(f"\n📊 분석 가능 항목:")
        print("   - 사용자별 이벤트 패턴 (성별, 학년, 학교별)")
        print("   - 디바이스/OS별 사용 패턴")
        print("   - 세션별 사용자 행동 분석")
        print("   - 학교/학급별 앱 사용 패턴")
        
        if analysis_result:
            print(f"\n⚠️ 데이터 특성:")
            print(f"   - 전체 user_id 중 {analysis_result['numeric_count']:,}개가 숫자 형태")
            print(f"   - {analysis_result['string_count']:,}개는 문자열 형태 (제외됨)")
            if analysis_result['mixed_count'] > 0:
                print(f"   - {analysis_result['mixed_count']:,}개는 혼합 형태 (제외됨)")
        
    else:
        print("❌ 조인 불가 - 2-way 분석만 가능")
        print("💡 대안: events+properties 또는 properties+users 개별 분석")

# 실행
if __name__ == "__main__":
    create_filtered_join_strategy()
````

````python
import pandas as pd
import gc

# 파일 경로들
paths = {
    'votes_accounts': "gs://sprintda05_final_project/votes/accounts_user.parquet",
    'hackle_users': "gs://sprintda05_final_project/hackle/user_properties.parquet",
    'hackle_props': "gs://sprintda05_final_project/hackle/hackle_properties.parquet", 
    'hackle_events': "gs://sprintda05_final_project/hackle/hackle_events.parquet"
}

def debug_merge_process():
    """merge 과정 단계별 디버깅"""
    
    print("🔍 4-way 조인 단계별 디버깅")
    print("="*40)
    
    try:
        # 1. 매칭 사용자 찾기 (작은 샘플로)
        print("📊 1. 매칭 사용자 찾기...")
        
        votes_df = pd.read_parquet(paths['votes_accounts'], columns=['id'])
        votes_ids = set(votes_df['id'].astype(str))
        del votes_df
        gc.collect()
        
        users_df = pd.read_parquet(paths['hackle_users'], columns=['user_id'])
        users_ids = set(users_df['user_id'].astype(str))
        del users_df
        gc.collect()
        
        matched_users = votes_ids & users_ids
        sample_users = list(matched_users)[:100]  # 더 작은 샘플로
        print(f"   샘플 사용자: {len(sample_users):,}명")
        
        # 2. votes 샘플 로드
        print("\n📊 2. votes 샘플 로드...")
        votes_full = pd.read_parquet(paths['votes_accounts'])
        votes_sample = votes_full[votes_full['id'].astype(str).isin(sample_users)].copy()
        votes_sample['id_str'] = votes_sample['id'].astype(str)
        
        print(f"   votes 컬럼: {list(votes_sample.columns)}")
        print(f"   votes 행수: {len(votes_sample):,}")
        
        del votes_full
        gc.collect()
        
        # 3. hackle_users 샘플 로드
        print("\n📊 3. hackle_users 샘플 로드...")
        users_full = pd.read_parquet(paths['hackle_users'])
        users_sample = users_full[users_full['user_id'].astype(str).isin(sample_users)].copy()
        users_sample['user_id_str'] = users_sample['user_id'].astype(str)
        
        print(f"   users 컬럼: {list(users_sample.columns)}")
        print(f"   users 행수: {len(users_sample):,}")
        
        del users_full
        gc.collect()
        
        # 4. Step 1: votes + users 조인
        print("\n📊 4. Step 1 조인...")
        step1 = votes_sample.merge(users_sample, left_on='id_str', right_on='user_id_str', how='inner')
        
        print(f"   Step 1 컬럼: {list(step1.columns)}")
        print(f"   Step 1 행수: {len(step1):,}")
        print(f"   Step 1 사용자수: {step1['id'].nunique():,}")
        
        # 5. hackle_props 로드
        print("\n📊 5. hackle_props 로드...")
        props_full = pd.read_parquet(paths['hackle_props'])
        
        # 샘플 사용자 + 숫자 user_id 필터링
        props_sample = props_full[
            (props_full['user_id'] != '') & 
            (props_full['user_id'].astype(str).str.isdigit()) &
            (props_full['user_id'].astype(str).isin(sample_users))
        ].copy()
        props_sample['user_id_str'] = props_sample['user_id'].astype(str)
        
        print(f"   props 컬럼: {list(props_sample.columns)}")
        print(f"   props 행수: {len(props_sample):,}")
        print(f"   props 사용자수: {props_sample['user_id'].nunique():,}")
        
        del props_full
        gc.collect()
        
        # 6. Step 2: + props 조인 (suffixes 명시)
        print("\n📊 6. Step 2 조인...")
        step2 = step1.merge(props_sample, on='user_id_str', how='inner', suffixes=('', '_props'))
        
        print(f"   Step 2 컬럼: {list(step2.columns)}")
        print(f"   Step 2 행수: {len(step2):,}")
        
        # id 컬럼 확인
        id_columns = [col for col in step2.columns if 'id' in col.lower()]
        print(f"   ID 관련 컬럼들: {id_columns}")
        
        if 'id' in step2.columns:
            print(f"   Step 2 사용자수: {step2['id'].nunique():,}")
        else:
            print("   ❌ 'id' 컬럼 없음!")
            # 대체 가능한 컬럼 찾기
            if 'id_str' in step2.columns:
                print(f"   Step 2 사용자수 (id_str): {step2['id_str'].nunique():,}")
            return False
        
        # 7. hackle_events 샘플 로드
        print("\n📊 7. hackle_events 로드...")
        
        # step2에서 나온 session_id들만 필터링
        available_sessions = step2['session_id'].dropna().unique()
        print(f"   조인 가능한 세션: {len(available_sessions):,}개")
        
        if len(available_sessions) == 0:
            print("   ❌ 조인 가능한 세션 없음!")
            return False
        
        # events에서 해당 세션들만 로드
        events_full = pd.read_parquet(paths['hackle_events'])
        events_sample = events_full[events_full['session_id'].isin(available_sessions)].copy()
        
        print(f"   events 컬럼: {list(events_sample.columns)}")
        print(f"   events 행수: {len(events_sample):,}")
        print(f"   events 세션수: {events_sample['session_id'].nunique():,}")
        
        del events_full
        gc.collect()
        
        # 8. Step 3: + events 최종 조인
        print("\n📊 8. Step 3 최종 조인...")
        step3 = step2.merge(events_sample, on='session_id', how='inner', suffixes=('', '_events'))
        
        print(f"   Step 3 컬럼: {list(step3.columns)}")
        print(f"   Step 3 행수: {len(step3):,}")
        
        if 'id' in step3.columns:
            final_users = step3['id'].nunique()
            print(f"   ✅ 최종 사용자수: {final_users:,}명")
            
            if final_users > 0:
                events_per_user = len(step3) / final_users
                print(f"   📊 사용자당 평균 이벤트: {events_per_user:.1f}개")
                
                # 주요 컬럼 확인
                key_cols = ['id', 'user_id_str', 'event_key', 'gender', 'school_id', 'ban_status']
                available = [col for col in key_cols if col in step3.columns]
                print(f"   🔧 분석 가능 컬럼: {available}")
                
                return True, final_users, len(step3)
        else:
            print("   ❌ 최종 단계에서도 'id' 컬럼 없음!")
            return False, 0, 0
        
    except Exception as e:
        print(f"❌ 디버깅 오류: {e}")
        import traceback
        traceback.print_exc()
        return False, 0, 0
    finally:
        # 메모리 정리
        try:
            del votes_sample, users_sample, props_sample, events_sample
            del step1, step2, step3
        except:
            pass
        gc.collect()

def calculate_final_estimate():
    """최종 예상치 계산"""
    
    print(f"\n📈 최종 예상치 계산...")
    
    success, final_users, final_rows = debug_merge_process()
    
    if success and final_users > 0:
        # 100명 샘플에서 성공한 비율
        sample_success_rate = final_users / 100
        
        # 전체 매칭된 사용자 기준 예상
        total_matched = 226329
        estimated_final = int(total_matched * sample_success_rate)
        
        # 전체 votes 대비 비율
        total_votes = 677085
        success_rate = (estimated_final / total_votes) * 100
        
        print(f"   샘플 성공률: {sample_success_rate*100:.1f}%")
        print(f"   📊 전체 예상 분석 가능 사용자: {estimated_final:,}명")
        print(f"   📊 전체 votes 대비: {success_rate:.1f}%")
        
        return estimated_final, success_rate, final_rows
    else:
        return 0, 0, 0

def provide_final_corrected_solution():
    """최종 수정된 해결책"""
    
    print(f"\n{'='*60}")
    print("🎯 최종 votes → hackle_events 분석 결과")
    print(f"{'='*60}")
    
    estimated_users, success_rate, sample_events = calculate_final_estimate()
    
    print(f"\n📊 최종 결과:")
    print(f"="*25)
    
    if estimated_users > 0:
        print(f"✅ 4-way 조인 성공!")
        print(f"🎯 예상 분석 가능 사용자: {estimated_users:,}명")
        print(f"📊 전체 votes 사용자 대비: {success_rate:.1f}%")
        print(f"📊 샘플 기준 예상 총 이벤트: {sample_events * (estimated_users/100):,.0f}행")
        
        print(f"\n🔧 수정된 최종 코드:")
        print("```python")
        print("import pandas as pd")
        print("")
        print("# 1) 매칭 사용자 찾기")
        print("votes_ids = set(pd.read_parquet(path_votes, columns=['id'])['id'].astype(str))")
        print("users_ids = set(pd.read_parquet(path_hackle_users, columns=['user_id'])['user_id'].astype(str))")
        print("matched_users = votes_ids & users_ids")
        print("")
        print("# 2) 매칭된 사용자만 로드")
        print("votes = pd.read_parquet(path_votes)")
        print("votes_matched = votes[votes['id'].astype(str).isin(matched_users)].copy()")
        print("votes_matched['id_str'] = votes_matched['id'].astype(str)")
        print("")
        print("users = pd.read_parquet(path_hackle_users)")
        print("users_matched = users[users['user_id'].astype(str).isin(matched_users)].copy()")
        print("users_matched['user_id_str'] = users_matched['user_id'].astype(str)")
        print("")
        print("# 3) Step 1: votes + users")
        print("step1 = votes_matched.merge(users_matched, left_on='id_str', right_on='user_id_str')")
        print("")
        print("# 4) Step 2: + props (suffixes 명시)")
        print("props = pd.read_parquet(path_hackle_props)")
        print("props_clean = props[(props['user_id'] != '') & (props['user_id'].astype(str).str.isdigit())].copy()")
        print("props_clean['user_id_str'] = props_clean['user_id'].astype(str)")
        print("step2 = step1.merge(props_clean, on='user_id_str', suffixes=('', '_props'))")
        print("")
        print("# 5) Step 3: + events (최종)")
        print("events = pd.read_parquet(path_hackle_events)")
        print("final = step2.merge(events, on='session_id', suffixes=('', '_events'))")
        print("")
        print(f"print(f'최종 사용자: {{final[\"id\"].nunique():,}}명')")
        print(f"print(f'총 이벤트: {{len(final):,}}행')")
        print("```")
        
        print(f"\n🚀 이 {estimated_users:,}명으로 가능한 분석:")
        print("   - 투표 참여자의 구체적 앱 사용 패턴")
        print("   - 투표 행동과 앱 이벤트 연관성")
        print("   - 학교별/성별 투표↔앱사용 차이")
        print("   - ban_status에 따른 행동 변화")
        print("   - 세션별 상세 이벤트 시퀀스 분석")
        
    else:
        print("❌ 4-way 조인 실패")
        print("💡 단계별 조인만 가능")
        print("🔍 위 디버깅 결과를 확인하여 문제점 파악")
    
    return estimated_users > 0

# 실행
if __name__ == "__main__":
    print("🚀 디버깅된 votes → hackle_events 분석")
    result = provide_final_corrected_solution()
    print(f"\n🧹 메모리 정리 완료")
    gc.collect()
````
