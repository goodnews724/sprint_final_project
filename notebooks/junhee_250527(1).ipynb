```python
import ast

# friend_id_list ì•ˆì „í•˜ê²Œ íŒŒì‹±
def safe_parse(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    elif isinstance(x, list):
        return x
    else:
        return []

accounts_user['friend_id_list'] = accounts_user['friend_id_list'].apply(safe_parse)

# ëª¨ë“  ì¹œêµ¬ ID ì¶”ì¶œ
all_friend_ids = set(
    friend_id
    for sublist in accounts_user['friend_id_list']
    for friend_id in sublist
)

# polls_usercandidateì˜ id ì§‘í•©
candidate_ids = set(polls_usercandidate['id'])

# ì¡´ì¬ ì—¬ë¶€ ë¹„êµ
missing_ids = all_friend_ids - candidate_ids

# ê²°ê³¼ ì¶œë ¥
if not missing_ids:
    print('âœ… ëª¨ë“  ì¹œêµ¬ IDê°€ polls_usercandidate.idì— ì¡´ì¬í•©ë‹ˆë‹¤.')
else:
    print(f'âš ï¸ ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ì¹œêµ¬ ID ìˆ˜: {len(missing_ids)}')
    print(f'ì˜ˆì‹œ: {list(missing_ids)[:10]}')
```
