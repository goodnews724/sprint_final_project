```python
import pandas as pd

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
user_df = pd.read_csv('/home/jovyan/work/data/accounts_user_processed.csv', parse_dates=['created_at'])
group_df = pd.read_csv('/home/jovyan/work/data/accounts_group.csv')
school_df = pd.read_csv('/home/jovyan/work/data/accounts_school.csv')
```

```python
# ìœ ì € í…Œì´ë¸” + ê·¸ë£¹ ì •ë³´ ì¡°ì¸ â†’ í•™êµ id ì–»ê¸°
user_group = pd.merge(user_df, group_df, left_on='group_id', right_on='id', how='left')
user_group = user_group.rename(columns={'school_id': 'school_id'})

# í•™êµë³„ë¡œ ê°€ì¥ ì²« ê°€ì… ì‹œì  ê³„ì‚°
school_first_join = (
    user_group
    .groupby('school_id')['created_at']
    .min()
    .reset_index()
    .rename(columns={'created_at': 'first_join_date'})
)

# í•™êµ ì´ë¦„, ì£¼ì†Œ ë“± ë¶™ì´ê¸°
school_info = pd.merge(school_first_join, school_df, left_on='school_id', right_on='id', how='left')

# í¼ì§„ ìˆœì„œëŒ€ë¡œ ì •ë ¬
school_info_sorted = school_info.sort_values('first_join_date').reset_index(drop=True)

# ê²°ê³¼ í™•ì¸
import datetime
school_info_sorted['first_join_date'] = school_info_sorted['first_join_date'].dt.strftime('%Y-%m-%d')
school_info_sorted[['first_join_date', 'school_id', 'address', 'school_type', 'student_count']].head(10)
```
