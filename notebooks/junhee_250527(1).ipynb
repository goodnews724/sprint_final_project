```python
import os
import pandas as pd
from google.cloud import storage

# GCS에서 Parquet 파일 불러오기 및 형식 확인
print("\n--- 2. GCS Parquet 파일 로드 및 형식 확인 ---")
gcs_parquet_path = "gs://sprintda05_final_project/votes/polls_usercandidate.parquet"

print(f"📂 다음 Parquet 파일을 불러옵니다: {gcs_parquet_path}")

# pandas를 사용하여 Parquet 파일 불러오기
# 'engine'을 'pyarrow'로 명시하여 확실하게 pyarrow 사용
# gcsfs가 설치되어 있다면 pandas는 gs:// 경로를 자동으로 처리
polls_usercandidate = pd.read_parquet(gcs_parquet_path, engine='pyarrow')

print("\n✅ Parquet 파일 로드 성공!")

print("\n--- 데이터프레임 상위 5개 행 (polls_usercandidate.head()) ---")
print(polls_usercandidate.head())

print("\n--- 데이터프레임 정보 (polls_usercandidate.info()) ---")
# 각 컬럼의 이름, Non-Null 개수, 데이터 타입(Dtype) 확인
polls_usercandidate.info()

print("\n--- 데이터프레임 기술 통계 (polls_usercandidate.describe()) ---")
# 숫자형 컬럼에 대한 통계 정보 확인
print(polls_usercandidate.describe(include='all')) # 모든 컬럼 타입 포함

print(f"\n--- 데이터프레임 행과 열 개수 (polls_usercandidate.shape): {polls_usercandidate.shape} ---")

print("\n--- 데이터프레임 컬럼 목록 (polls_usercandidate.columns) ---")
print(polls_usercandidate.columns.tolist())
```

```python
import ast

# friend_id_list 안전하게 파싱
def safe_parse(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    elif isinstance(x, list):
        return x
    else:
        return []

accounts_user['friend_id_list'] = accounts_user['friend_id_list'].apply(safe_parse)

# 모든 친구 ID 추출
all_friend_ids = set(
    friend_id
    for sublist in accounts_user['friend_id_list']
    for friend_id in sublist
)

# polls_usercandidate의 id 집합
candidate_ids = set(polls_usercandidate['id'])

# 존재 여부 비교
missing_ids = all_friend_ids - candidate_ids

# 결과 출력
if not missing_ids:
    print('✅ 모든 친구 ID가 polls_usercandidate.id에 존재합니다.')
else:
    print(f'⚠️ 존재하지 않는 친구 ID 수: {len(missing_ids)}')
    print(f'예시: {list(missing_ids)[:10]}')
```

```python
existing_ids = all_friend_ids & candidate_ids
print(list(existing_ids)[:30])  # 앞에서 30개만 보기

print(f'✅ 존재하는 친구 ID 수: {len(existing_ids)}')

```

```python
subset = pd.DataFrame()  # 항상 초기화

# viewer_id 지정
viewer_id = 831958

# 친구 목록 가져오기 (없으면 빈 set)
try:
    friend_list = set(accounts_user.loc[accounts_user['id'] == viewer_id, 'friend_id_list'].values[0])
except IndexError:
    print(f'viewer_id {viewer_id} not found in accounts_user')
    friend_list = set()

# viewer가 받은 질문 ID 추출 (없으면 빈 array)
viewer_questions = polls_usercandidate.loc[
    polls_usercandidate['user_id'] == viewer_id, 'question_piece_id'
].unique()

# 받은 질문이 없으면 이후 로직은 스킵
if len(viewer_questions) == 0:
    print(f'viewer_id {viewer_id}는 받은 질문이 없습니다.')
    subset = pd.DataFrame()  # 빈 데이터프레임 생성
else:
    # 해당 질문들에 나오는 후보만 추출
    subset = polls_usercandidate[polls_usercandidate['question_piece_id'].isin(viewer_questions)].copy()

    # 친구 여부 판단
    subset['is_friend_with_viewer'] = subset['user_id'].isin(friend_list)

# 결과 확인 (subset이 비었는지 먼저 체크)
if subset.empty:
    print(f'viewer_id {viewer_id}에 해당하는 후보자 데이터가 없습니다.')
else:
    print(subset[['question_piece_id', 'user_id', 'is_friend_with_viewer']].head())
    print(subset['is_friend_with_viewer'].value_counts())
    print('고유 후보자 수:', subset['user_id'].nunique())
```
