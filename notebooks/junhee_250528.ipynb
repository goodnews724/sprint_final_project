```python
import pandas as pd

# í…Œì´ë¸” ì´ë¦„ ë¦¬ìŠ¤íŠ¸
table_names = [
    'accounts_attendance',
    'accounts_blockrecord',
    'accounts_failpaymenthistory',
    'accounts_friendrequest',
    'accounts_group',
    'accounts_nearbyschool',
    'accounts_paymenthistory',
    'accounts_user_contacts',
    'accounts_pointhistory',
    'accounts_school',
    'accounts_timelinereport',
    'accounts_user',
    'accounts_userquestionrecord',
    'accounts_userwithdraw',
    'event_receipts',
    'events',
    'polls_question',
    'polls_questionpiece',
    'polls_questionreport',
    'polls_questionset',
    'polls_usercandidate'
]

# GCS ê²½ë¡œ ë”•ì…”ë„ˆë¦¬ ìƒì„±
gcs_parquet_paths = {
    name: f'gs://sprintda05_final_project/votes/{name}.parquet' for name in table_names
}

# ê²°ê³¼ ì €ì¥ìš© ë”•ì…”ë„ˆë¦¬
dfs = {}

# ë°˜ë³µë¬¸ìœ¼ë¡œ ê° íŒŒì¼ ë¡œë”© ë° ì •ë³´ ì¶œë ¥
for name, path in gcs_parquet_paths.items():
    print(f"\nğŸ“‚ --- {name} ë¡œë“œ ì‹œì‘ ---")
    print(f"íŒŒì¼ ê²½ë¡œ: {path}")
    
    try:
        df = pd.read_parquet(path, engine='pyarrow')
        dfs[name] = df  # ë”•ì…”ë„ˆë¦¬ì— ì €ì¥

        print("âœ… ë¡œë“œ ì„±ê³µ!")
        print(f"ğŸ”¹ ìƒìœ„ 5ê°œ í–‰ ({name})")
        print(df.head())
        print(f"\nğŸ”¹ ì»¬ëŸ¼ ëª©ë¡: {df.columns.tolist()}")
        print(f"ğŸ”¹ í–‰ê³¼ ì—´ ê°œìˆ˜: {df.shape}")
        print(f"ğŸ”¹ ì •ë³´ ìš”ì•½:")
        df.info()
        print("\n" + "-"*60)
    except Exception as e:
        print(f"âŒ {name} ë¡œë“œ ì‹¤íŒ¨: {e}")
        print("\n" + "-"*60)
```

```python
def check_missing_duplicates_outliers(df, z_thresh=3):
    import numpy as np

    print("ğŸ“Œ [1] ê²°ì¸¡ì¹˜(Missing Values)")
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        print(missing.sort_values(ascending=False))
    else:
        print("âœ… ê²°ì¸¡ì¹˜ ì—†ìŒ")

    print("\nğŸ“Œ [2] ì¤‘ë³µê°’(Duplicates)")
    dup_count = df.duplicated().sum()
    print(f"ì´ ì¤‘ë³µ í–‰ ìˆ˜: {dup_count}")

    print("\nğŸ“Œ [3] ì´ìƒì¹˜(Outliers, Z-score ê¸°ì¤€)")
    numeric_cols = df.select_dtypes(include=[np.number])
    if numeric_cols.empty:
        print("âš ï¸ ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ ì—†ìŒ â†’ ì´ìƒì¹˜ í™•ì¸ ë¶ˆê°€")
    else:
        from scipy.stats import zscore
        z_scores = np.abs(zscore(numeric_cols, nan_policy='omit'))
        outlier_mask = (z_scores > z_thresh).any(axis=1)
        outlier_count = outlier_mask.sum()
        print(f"{z_thresh} ì´ìƒ Z-score ê¸°ì¤€ ì´ìƒì¹˜ í–‰ ìˆ˜: {outlier_count}")
```
